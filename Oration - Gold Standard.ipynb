```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Get list of files and folders in Adgooroo directory
path = r'/Users/smacrae/Documents/gold_standard/Table'
folderlist = os.listdir(path)

# Remove .DS_Store from folderlist
folderlist = folderlist[1:]

# Folderlist header
folderlist[0:10]
```

```python
data = []
for dir_entry in folderlist:
    dir_entry_path = os.path.join(path, dir_entry)
    if os.path.isfile(dir_entry_path):
        data.append(dir_entry_path)
        
data[0:10]
```

```python
# Write each table into a list of pandas dataframes
pieces = []

for i in range(0, len(data)):
    table = pd.read_csv(data[i], quoting=0).dropna(axis=1,how='all')
    # remove utf-8-sig encoding from column name and re-encode as utf-8
    new_cols = []
    for j in table.columns:
        new_cols.append(j.replace('\"', '').strip().decode("utf-8-sig").encode("utf-8"))
    # update column names
    table.columns = new_cols
    # append to list
    pieces.append(table)
```

```python
# RxNorm
file_name = "Marketed_Product_RxNorm.csv"
rxnorm = pieces[pd.Series(folderlist)[pd.Series(folderlist) == file_name].index[0]]
pd.Series(rxnorm.columns).tolist()
```

```python
# Product
file_name = "Product.csv"
product = pieces[pd.Series(folderlist)[pd.Series(folderlist) == file_name].index[0]]
pd.Series(product.columns).tolist()
```

```python
# Product Strength
file_name = "Product_Strength_Route_Form.csv"
product_strength = pieces[pd.Series(folderlist)[pd.Series(folderlist) == file_name].index[0]]
pd.Series(product_strength.columns).tolist()
```

```python
# Product Indications
file_name = "Product_Indications.csv"
product_indications = pieces[pd.Series(folderlist)[pd.Series(folderlist) == file_name].index[0]]
pd.Series(product_indications.columns).tolist()
```

```python
# Package
file_name = "Package.csv"
package = pieces[pd.Series(folderlist)[pd.Series(folderlist) == file_name].index[0]]
pd.Series(package.columns).tolist()
```

```python
# Price
file_name = "Price.csv"
price = pieces[pd.Series(folderlist)[pd.Series(folderlist) == file_name].index[0]]
pd.Series(price.columns).tolist()
```

```python
print 'rxnorm:', '\n', len(rxnorm)
print 'product:', '\n', len(product)
print 'product_strength:', '\n', len(product_strength)
print 'product_indications:', '\n', len(product_indications)
print 'package:', '\n', len(package)
print 'price:', '\n', len(price)
```

```python
# Merge RxNorm and Product tables
df = pd.merge(rxnorm, product, on='MarketedProductID', how='left', suffixes=['_rxnorm','_product'])
print 'merge 1 (rxnorm + product):', '\n', len(df)
df = pd.merge(df, product_strength, on='ProductID', how='left', suffixes=['_df','_strength'])
print 'merge 2 (... + product_strength):', '\n', len(df)
# df = pd.merge(df, product_indications, on='ProductID', how='left', suffixes=['_df','_strength'])
# print 'merge 3 (... + product_indications):', '\n', len(df)
df = pd.merge(df, package, on='ProductID', how='left', suffixes=['_df','_package'])
print 'merge 4 (... + package):', '\n', len(df)
# df = pd.merge(df, price, on='PackageID', how='left', suffixes=['_df','_package'])
```
