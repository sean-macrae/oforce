```python
# Calculate last overall fill date
df['last_fill_date'] = pd.to_datetime(df['service_dt']).max()
pd.to_datetime(df['service_dt']).max()
```

```python
df['service_dt'] = pd.to_datetime(df['service_dt'])
```

```python
# Filter days supply >20 and <365
df = df.loc[(df['days_supply'] > 20) & (df['days_supply'] < 365),].reset_index(drop=True)

# Filter quantity > 0
df = df.loc[df['quantity_df'] > 0,].reset_index(drop=True)

# Filter for Capsules & Tablets only
# df = df.loc[(df['dose_form_df'] == 'Oral Capsule') | (df['dose_form_df'] == 'Oral Tablet'), ].reset_index(drop=True)
```

```python
# 1) Calculate active window
df['active_window'] = 0

df.loc[df['days_supply'] <= 31,'active_window'] = pd.Timedelta('90 days')
df.loc[(df['days_supply'] > 31) & (df['days_supply'] <= 91),'active_window'] = pd.Timedelta('180 days')
df.loc[df['days_supply'] > 91,'active_window'] = pd.Timedelta('360 days')

df['active_date'] = pd.to_datetime(df['last_fill_date']) - df['active_window']

# Service date >= Active date
df['rule_1'] = pd.Series(pd.to_datetime(df['active_date']) <= pd.to_datetime(df['service_dt']))
```

```python
# 2) Calculate next service date window
df_last_fill = df[['primary_oration_id','oration_id','ndc',
                   'service_dt']].groupby([
                   'primary_oration_id','oration_id','ndc']).max().reset_index(drop=False)

df_last_fill.columns = ['primary_oration_id','oration_id','ndc','service_dt_max']

df = pd.merge(df, df_last_fill, on=['primary_oration_id','oration_id','ndc'])
```

```python
# Calculate next fill date
fill_window = []

for i in range(0, len(df)):
    fill_window.append(pd.Timedelta(str(df['days_supply'][i])+' days'))

df['fill_window'] = fill_window

for i in range(0, len(df)):
    df['fill_date'] = pd.to_datetime(df['service_dt_max'][i]) + df['fill_window']

df['fill_date'] = pd.to_datetime(df['service_dt']) + df['fill_window']

df['rule_2'] = pd.Series(df['fill_date'] >= df['last_fill_date'] - pd.Timedelta(str('45 days')))
```

```python
df = df.loc[(df['rule_1'] == True) & (df['rule_2'] == True), ]
```
