```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import os
```

```python
# get list of files in log level directory
logpath = r'/Users/smacrae/Documents/data_warehouse/log/EventLogData-utilservice.access.20131112'
loglist = os.listdir(logpath)

# # get list of files in optimization directory
# optpath = r'/Users/smacrae/Documents/data_warehouse/opt'
# optlist = os.listdir(optpath)
```

```python
# create list of strings for the directory path of each files in Adgooroo directory
logdata = []
for dir_entry in os.listdir(logpath):
    dir_entry_path = os.path.join(logpath, dir_entry)
    if os.path.isfile(dir_entry_path):
        logdata.append(dir_entry_path)
```

```python
# write each log file into a list of pandas dataframes
pieces = []

for i in range(0, len(logdata)):
    log = pd.read_csv(logdata[i], sep='\t', header=None, index_col=None, parse_dates=True, infer_datetime_format=True)    
    log['directory'] = logdata[i]
    pieces.append(log)
```

```python
# combine list of dataframes into a single dataframe
logs = pd.concat(pieces, ignore_index=True)
```

```python
logs.head()
```

```python
# parse column 0 between timestamps and IP address
temp_0 = pd.DataFrame(logs[0].str.split(',').tolist())
temp_1 = pd.DataFrame(temp_0[1].str.split(' - ').tolist())
temp_1.columns = ['id','ip']

temp = temp_0.join(temp_1)[[0, 'id', 'ip']]
temp.columns = ['datetime','id','ip']
```

```python
logs = logs[[1, 'directory']].join(temp)
logs.columns = ['log','directory','datetime','id','ip']
```

```python
events = pd.DataFrame(logs['log'].str.split('/').tolist())
events.columns = ['blank','api','version','type','external_member_id','event']
```

```python
actions = events['event'].str.split('=').tolist()

temp = []
for i in range(0,len(actions)):
    if actions[i] == None:
        temp.append(['Null'])
    else:
        temp.append(actions[i])
        
actions = pd.DataFrame(temp)
actions.columns = ['event_type','event_name']
```

```python
events = events.join(actions)[['api','version','type','external_member_id','event_type','event_name']]
```

```python
logs = logs.join(events)[['external_member_id','id','datetime','event_type','event_name','type','version','api','ip','directory']]
logs.describe()
```

```python
# append datetime meta-attributes
logs['year'] = pd.DatetimeIndex(pd.to_datetime(logs['datetime'])).year
logs['month'] = pd.DatetimeIndex(pd.to_datetime(logs['datetime'])).month
logs['day'] = pd.DatetimeIndex(pd.to_datetime(logs['datetime'])).day
logs['weekday'] = pd.DatetimeIndex(pd.to_datetime(logs['datetime'])).weekday
logs['hour'] = pd.DatetimeIndex(pd.to_datetime(logs['datetime'])).hour
logs['time'] = pd.DatetimeIndex(pd.to_datetime(logs['datetime'])).time
```

```python
logs = logs[['external_member_id','id','datetime','year','month','day','weekday','hour','time','event_type','event_name','type','version','api','ip','directory']]
logs.head()
```

```python
logs.to_csv('/Users/smacrae/Documents/data_warehouse/log/combined_logs.csv', index=False)
```
