```python
import numpy as np
import pandas as pd
from pandas.io import gbq
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

```python
#####################################################################
# Load Data
#####################################################################
```

```python
###############################################################################
# Query AdWords.ad from Google BigQuery API

projectid = 'aerial-jigsaw-823'

logs_query = """SELECT
log.datetime AS datetime,
log.external_member_id AS external_member_id,
ees.oration_id AS oration_id,
ees.primary_oration_id AS primary_oration_id,
ees.org_name AS org_name,
ees.yob AS yob,
ees.dependents AS dependents,
ees.relationship AS relationship,
log.event_name AS event_name
FROM
log.detail log
LEFT JOIN
(SELECT
primary_oration_id AS primary_oration_id,
oration_id AS oration_id,
member_id AS external_member_id,
org_short_name AS org_name,
date_part AS yob,
cnt AS dependents,
relationship AS relationship
FROM
ees.enrollment
GROUP BY
primary_oration_id, oration_id, external_member_id, org_name, yob, dependents, relationship) ees
ON
log.external_member_id = ees.external_member_id
WHERE 
org_name NOT IN ('ORG0006','ORG0007') AND log.event_name IS NOT NULL
GROUP BY
datetime, external_member_id, oration_id, primary_oration_id, org_name, yob, dependents, relationship, event_name"""

opts_query = """SELECT *
FROM
opt.summary
WHERE
org_name NOT IN ('ORG0006','ORG0007');"""

logs = pd.read_gbq(logs_query, project_id = projectid, reauth=False)
opts = pd.read_gbq(opts_query, project_id = projectid, reauth=False)
```

```python
# Pre-process optimization data
```

```python
# Map lf_chains < 2% of observations as 'other'
lf_chain = pd.DataFrame(opts.groupby(['lf_chain']).apply(len)).sort(columns=0, ascending=False).reset_index()
lf_chain.columns = ['lf_chain','frequency']

lf_chain['frequency_pct'] = lf_chain['frequency'] / lf_chain['frequency'].sum()
lf_chain

lf_chain['map_bool'] = lf_chain['frequency_pct'] > 0.03
lf_chain

map_bool = []

for i in range(0, len(lf_chain)):
    if lf_chain.loc[i, 'map_bool'] == True:
        map_bool.append(lf_chain.loc[i, 'lf_chain'])
    else:
        map_bool.append(u'Other_Chain')
        
lf_chain['lf_chain_map'] = map_bool
lf_chain

opts = pd.merge(opts, lf_chain[['lf_chain','lf_chain_map']], on='lf_chain', how='left')

# Create dummy columns for lf_chain_map
opts = pd.concat([opts, pd.get_dummies(opts['lf_chain_map'])], axis=1)

lf_chain_len = len(lf_chain.loc[lf_chain['map_bool'] == True,'lf_chain_map'])
```

```python
# Number of optimizations
opts_len = opts.groupby(['org_name','primary_oration_id','oration_id'])[['custom_name']].apply(len).reset_index()
opts_len.columns = ['org_name','primary_oration_id','oration_id','optimizations']
```

```python
list(opts.columns)[(len(opts.columns)-(lf_chain_len)-1):len(opts.columns)]
```

```python
# Average rebate offer, fills per year, member savings, organization savings
opts_avg = opts.groupby(['org_name','primary_oration_id','oration_id'])[[
                'rebate','lf_fills_per_year','member_annual_savings','org_annual_savings',
                'CVS','Express Scripts Home Delivery','Other_Chain','Rite Aid',
                'Safeway','Target','Walgreens','Walmart']].apply(sum).reset_index()

opts = pd.merge(opts_len, opts_avg)
opts.head()
```

```python
# Pre-process log data
```

```python
# Create df from logs for survival analysis

# Start time
first_obs = logs.groupby(['primary_oration_id','oration_id','org_name','yob','dependents','relationship'])[['datetime']].apply(min).reset_index()
first_obs.columns = ['primary_oration_id','oration_id','org_name','yob','dependents','relationship','start']

# Event time (registration complete)
first_obs_event = logs.groupby(['primary_oration_id','oration_id','event_name'])[['datetime']].apply(min).reset_index()
first_obs_event.columns = ['primary_oration_id','oration_id','event','time']
```

```python
# Calculate difference between start and event observations
df = pd.merge(first_obs, first_obs_event, on=['primary_oration_id','oration_id'], how='left')
df['delta'] = df['time'] - df['start']

# Encode columns T (time) and E (whether event is present or not)
df['T'] = (pd.DatetimeIndex(df['delta']).day-1)*24 + pd.DatetimeIndex(df['delta']).hour
df['E'] = pd.Series(df['event'] == 'registration_complete').astype(int)

# Calculate age
df['age'] = 2015 - df['yob']

# Create dummy variables for org
df = pd.concat([df, pd.get_dummies(df['org_name'])], axis=1)

# Create dummy variables for relationship
df = pd.concat([df, pd.get_dummies(df['relationship'])], axis=1)

df.head()
```

```python
# Merge optimization summary stats with log level detail
df = pd.merge(df, opts, on = ['primary_oration_id','oration_id','org_name'], how='left')
df.head()
```

```python
# Remove external member ID, org, relationship from df
df = df[df.columns - ['primary_oration_id','oration_id','relationship','event','yob','start','time','delta']]

df.head()
```

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
from lifelines import KaplanMeierFitter

T = df['T']
E = df['E']

kmf = KaplanMeierFitter()
kmf.fit(T, event_observed=E)
```

```python
# Kaplan-meier Survival Function

figsize(12.5, 7.5)

kmf.plot()
```

```python
groups = df['org_name']
ix = groups == 'YMKT'

kmf.fit(T[ix], E[ix], label='YMKT')
ax = kmf.plot()

kmf.fit(T[~ix], E[~ix], label='MAF')
kmf.plot(ax=ax)
```

```python
# Cumulative Hazard Function

from lifelines import NelsonAalenFitter
naf = NelsonAalenFitter()
naf.fit(T, event_observed=E)

ax = naf.plot()

naf.fit(T[~ix], E[~ix], label='MAF')
naf.plot(ax=ax)
```

```python
# Survival Regression

# While the above KaplanMeierFitter and NelsonAalenFitter are useful, 
# they only give us an “average” view of the population.

# Often we have specific data at the individual level,
# either continuous or categorical, that we would like to use.
# For this, we turn to survival regression, specifically AalenAdditiveFitter or CoxPHFitter.
```

```python
df = df[df.columns - ['org_name','optimizations','MAF','YMKT','PTSN','Primary','Child','Spouse']]

df = df.dropna()
```

```python
df.describe()
```

```python
from lifelines import AalenAdditiveFitter, CoxPHFitter

# Using Cox Proportional Hazards model
cf = CoxPHFitter()
cf.fit(df, duration_col='T', event_col='E')
print cf.summary

# Using Aalen's Additive model
aaf = AalenAdditiveFitter(fit_intercept=False)
aaf.fit(df, duration_col='T', event_col='E')
```

```python
x = df[df.columns - ['E','T_hours']]
aaf.predict_survival_function(x.ix[10:12]).plot() #get the unique survival functions of the first two subjects
```
