```python
# Objective: Are prices legit?

# A: Can the avaialble data support our objective? No; it is sampled (small, biased), missing signals (unknowns).

# Then what can we do with available data...?

# A: We can estimate best in class gross_cost per unit (by rxcui) for this population, and this population alone, (and based on historical data) 
# ... and present it within the context of some confidence heuristic (based on frequency and variance).

# Meat & Potatoes -
# Definition: Best-in class gross_cost per unit = min(gross_cost per unit, excluding outliers)
# Inner-quartile range = Q3 - Q1
# Mild lower outlier: Q1 - 1.5*IQ (best in-class gross_cost per unit)
# Extreme lower outier: Q1 - 3*IQ (best in-class gross_cost per unit; aggressive)

# Observations -
# 1) If variance = 0, then best in class = mean.
# 2) If gross_cost is small, savings is likely also small (so highlight on RxCUI's with moderate - large gross_cost)
# 3) As frequency -> infinity, confidence in best in class estimate increases (so highlight RxCUI's with larger frequencies)

# Approach - 
# 1) Calculate outlier fences, variance, and frequency of gross cost per unit by RxCUI
# 2) If variance = 0, set to NaN
# 3) Sort by frequency, variance (to help sort assign cost, frequency tiers of 1 - 10)
# ... sort(cost-tier(desc), frequency-tier(desc), variance(asc))

# Conclusion -
# We now have a table of estimated best in class (moderate & aggressive) gross costs (for this sampled population)...
# ... sorted by savings opportunity and (relative) confidence.

# Deliverable - 
# 1) Best in class gross cost table (maybe sample of boxplot graph with illustrative RxCUI's; some different shape)

# Boxplot
df_box = df[df['cost_per'] < 0.05]
df_box.boxplot(column=['cost_per'], by=['rxcui'])
```

```python
# create dictionary of rxcui and cost_per
rx_cost = df[['rxcui','cost_per']]
rx_cost = {k: g['cost_per'].tolist() for k,g in rx_cost.groupby('rxcui')}
```

```python
#############################################################################################################################
### Descriptive Summary by RxCUI
#############################################################################################################################
```

```python
# define function to calculate summary stats over dataframe
def g(x):
    return pd.Series([len(x.index),x.min(),x.max(),x.mean(),x.median(),x.std(),x.quantile(q=0.05),x.quantile(q=0.95)], index=['Records','Min','Max','Mean','Median','StdDev','Lower5th','Upper5th'])
```

```python
# group dataframe by Advertiser and apply summary stats to their Coverage
rxcui_groups = df.groupby(df['rxcui'])
rxcui_1 = rxcui_groups['quantity'].apply(g).unstack()
rxcui_1.columns = ['Qty_Records','Qty_Min','Qty_Max','Qty_Mean','Qty_Median','Qty_StdDev','Qty_Lower5th','Qty_Upper5th']

rxcui_2 = rxcui_groups['gross_cost'].apply(g).unstack()
rxcui_2.columns = ['Cost_Records','Cost_Min','Cost_Max','Cost_Mean','Cost_Median','Cost_StdDev','Cost_Lower5th','Cost_Upper5th']

rxcui_3 = rxcui_groups['cost_per'].apply(g).unstack()
rxcui_3.columns = ['CostPer_Records','CostPer_Min','CostPer_Max','CostPer_Mean','CostPer_Median','CostPer_StdDev','CostPer_Lower5th','CostPer_Upper5th']
```

```python
rxcui_stats = pd.merge(rxcui_1, rxcui_2, left_index=True, right_index=True)
rxcui_stats = pd.merge(rxcui_stats, rxcui_3, left_index=True, right_index=True)
rxcui_stats = rxcui_stats.sort(['Qty_Records'], ascending=False)
rxcui_stats.head()
```

```python
rxcui_stats = rxcui_stats[rxcui_stats['CostPer_Median'] < 1000]
rxcui_stats = rxcui_stats[rxcui_stats['Qty_Records'] > rxcui_stats['Qty_Records'].quantile(q=0.95)]
# rxcui_stats = rxcui_stats.head()
```

```python
import matplotlib.pyplot as plt

N = len(rxcui_stats)
ypos = np.arange(N)

plt.figure(figsize=(8, 25))
plt.barh(ypos, rxcui_stats['CostPer_Median'], xerr=rxcui_stats['CostPer_StdDev'], align='center', alpha=0.001, color='blue')
plt.plot(rxcui_stats['CostPer_Median'], ypos, marker='o', markersize=3, color='black', ls='')
# plt.plot(rxcui_stats['CostPer_Mean'], ypos, marker='o', markersize=3, color='black', ls='')

plt.yticks(ypos, rxcui_stats.index)
plt.title('Distribution Gross Cost Per Unit')
plt.xlabel('Gross Cost Per Unit')
plt.ylabel('RxCUI')

plt.show()
```

```python
x = pd.DataFrame(rx_cost[1232088])
```

```python
# specify normal distribution
center = pm.Normal("center", x.mean(), 1/(x.var()))

# set prior
p = pm.Uniform('p', lower=0, upper=x.max())
```

```python
print "Random output:", center.random(), center.random(), center.random()
```

```python
@pm.deterministic
def lambda_(center=center):
    return center

model = pm.Model([lambda_])
```

```python
# run simulation
mcmc = pm.MCMC(model)
mcmc.sample(50000, burn=5000)
```

```python
figsize(12.5, 4)
plt.title("Posterior distribution of $p_A$, the cost of $R_x$ drug")
plt.hist(mcmc.trace('lambda_')[:], histtype="stepfilled", normed=True)
plt.legend()
```
