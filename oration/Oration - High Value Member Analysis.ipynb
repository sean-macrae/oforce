```python
import numpy as np
import pandas as pd
from pandas.io import gbq
import matplotlib.pyplot as plt
```

```python
# Load optimization file as 'opts'
projectid = 'aerial-jigsaw-823'
opt_query = """SELECT * FROM opt.summary"""
    
# Optimization file (to identify converted members)
opts = pd.read_gbq(opt_query, project_id = projectid, reauth=False)
```

```python
pd.Series(opts.columns)
```

```python
### Pre-process data
# Filter for YMKT
df = opts.loc[opts['org_name'] == 'YMKT',['org_name','plan_number',
                                     'member_annual_savings','org_annual_savings']]
# Calculate total savings
df['total_savings'] = df['member_annual_savings'] + df['org_annual_savings']

# Filter records where savings < $1.00
df = df.loc[df['total_savings'] > 1.0,]

# Sort df by total_savings and reset_index
df = df.sort(columns=['total_savings'], ascending=False).reset_index(drop=True)

# Calculate total savings cumulative sum
df['total_savings_cumsum'] = df['total_savings'].cumsum()

# Calculate percent of cumulative total savings
df['pct_total_savings'] = df['total_savings_cumsum']/df['total_savings'].sum()

# Calculate 80/20
df['80/20'] = df['pct_total_savings'] <= 0.80

# Subset df on members accounting for 80% of totall savings
df_80 = df.loc[df['80/20'] == True,['total_savings']]
```

```python
%matplotlib inline
df['total_savings'].plot(kind='kde', color='red', figsize=(12, 8),
                         xlim=(0,df['total_savings'].max()), 
                         title='Distribution of Total Savings')
df['total_savings'].hist(grid=False, bins=100, color='lightgrey', normed=True)
plt.show()

print 'Distribution of Total Savings:'
print pd.DataFrame(df['total_savings'].describe())
print '\n'
print '80% of total savings is associated with the upper ' + str((df.loc[df['80/20'] == True,'total_savings'].describe()['count'] / 
       df['total_savings'].describe()['count'] * 100).astype(int)) + '% of members'
```

```python
### Boxplot of 80% group
# Define inner-quartile range multiplier for calculating whisker of boxplot

whisker = 3.0

# Plot boxplot of 80/20 total savings group
df_80['total_savings'].plot(kind='box', notch=True, figsize=(12, 8), 
                            whis=whisker, sym = 'o',
                            title='80/20 Group with '+str(whisker)+'x Inner-quartile Whiskers')
plt.show();

# Calculate upper whisker of boxplot
whisker_upper = (df_80['total_savings'].quantile(0.75) + 
(df_80['total_savings'].quantile(0.75) - 
 df_80['total_savings'].quantile(0.25)) * whisker)

# Identify 80/20 outliers in df
df['80/20_outlier'] = df['total_savings'] > whisker_upper

print 'The upper-whisker is defined at ' + str(whisker_upper), '\n'

print str((df.loc[df['80/20_outlier'] == True,'total_savings'].describe()['count'] / 
df['total_savings'].describe()['count'] * 100).astype(int))+'% of members have been flagged as outliers', '\n'

print 'Outlier Distribution of Total Savings'
print pd.DataFrame(df.loc[df['80/20_outlier'] == True,'total_savings'].describe()), '\n'
```

```python
df_scatter = df

df_scatter.loc[df_scatter['member_annual_savings'] < 1,'member_annual_savings'] = 1
df_scatter.loc[df_scatter['org_annual_savings'] < 1,'org_annual_savings'] = 1

# Define color list for scatter plot
color_list = []

for i in range(0, len(df_scatter)):
    if df_scatter.loc[i, '80/20_outlier'] == True:
        color_list.append('r')
    elif df_scatter.loc[i, '80/20'] == True:
        color_list.append('y')
    else:
        color_list.append('b')
        

df_scatter['color'] = color_list
```

```python
df_scatter['plan_number'].unique()
```

```python
df_scatter.plot(kind='scatter',
        x='member_annual_savings',y='org_annual_savings',
        figsize=(12, 8), logx=True, logy=True,
        c = df_scatter['color'], 
        s=  df_scatter['total_savings'], alpha = 0.3,
        title='Member Savings vs. Org Savings')

plt.axvline(x=df['member_annual_savings'].median(), color='blue')
plt.axhline(y=df['org_annual_savings'].median(), color='blue')

plt.show()
```
