```python
# create GBRT test and training sets

# define training and test set ranges
# train dates
train_date_start = df_top['service_date'].max() - datetime.timedelta(days=(364+30))
train_date_end = df_top['service_date'].max() - datetime.timedelta(days=(30))
# test dates
test_date_start = df_top['service_date'].max() - datetime.timedelta(days=(30-1))
test_date_end = df_top['service_date'].max()
```

```python
# (GBRT) define metric signals

# GBRT signals
signals = ['BrandGenericStatusID',
           'PackageSize',
           'Strength',
           'PerVolume',
            'gender_factor',
            'is_brandname_factor',
            'fulfillment_factor',
            'claim_type_factor',
            'pbm_factor',
            'pharmacy_state_factor',
            'chain_factor',
            'employer_description_factor',
            'RxNormType_factor',
            'IngredientName_factor',
            'FDAForm_factor',
            'GSForm_factor',
            'TopLevelDoseFormName_factor',
            'date_year',
            'date_quarter',
            'date_month',
            'date_week']

target = ['gross_cost']

date = ['service_date']

gbrt_columns = date + signals + target

print 'Target:', len(target)
print target, '\n'

print 'Signals:', len(signals)
print signals, '\n'

print 'GBRT Columns:', len(gbrt_columns)
print gbrt_columns, '\n'
```

```python
df_gbrt = df_top[gbrt_columns]

# replace infinity with NaN
df_gbrt = df_gbrt.replace([np.inf, -np.inf], np.nan)

# remove NaN's
df_gbrt = df_gbrt.dropna()
```

```python
# gross_cost training
X_train_gbrt = df_gbrt[(df_gbrt['service_date'] >= train_date_start) & (df_gbrt['service_date'] <= train_date_end)][signals]
y_train_gbrt = df_gbrt[(df_gbrt['service_date'] >= train_date_start) & (df_gbrt['service_date'] <= train_date_end)][target[0]]
# gross_cost test
X_test_gbrt = df_gbrt[(df_gbrt['service_date'] >= test_date_start) & (df_gbrt['service_date'] <= test_date_end)][signals]
y_test_gbrt = df_gbrt[(df_gbrt['service_date'] >= test_date_start) & (df_gbrt['service_date'] <= test_date_end)][target[0]]
```

```python
# define model parameters
params_gbrt = {'n_estimators': 500, 'max_depth': len(signals), 'min_samples_split': 25, 'subsample': 0.70,
          'learning_rate': 0.01, 'loss': 'lad', 'verbose': 1}
```

```python
# training

# gross_cost
gbm_gbrt = ensemble.GradientBoostingRegressor(**params_gbrt)
gbm_gbrt.fit(X_train_gbrt, y_train_gbrt)
gbm_gbrt_mse = mean_squared_error(y_test_gbrt, gbm_gbrt.predict(X_test_gbrt))
gbm_gbrt_rsquared = np.corrcoef(y_test_gbrt, gbm_gbrt.predict(X_test_gbrt))[0,1]**2
print("gross_cost R-squared: %.4f" % gbm_gbrt_rsquared)
```

```python
# set index to service date (for time series analysis)
df_ols = df_top.set_index('service_date', drop=False)
df_ols['cpq']

# define columns for groupby and aggregations
df_ols_groups = df_ols.groupby(['rxcui'])

# define resampling function that aggregates to the month
def resample(x):
    return x.resample('D').resample('M', kind='period')

# resample RxCUI at monthly frequency
df_ols_monthly = pd.DataFrame(df_ols_groups['cpq'].apply(resample).unstack().T)
df_ols_monthly.head()
```

```python
import statsmodels.api as sm

# inititalize lists to store model parameters

# forecast_points
forecast_points = []

# rx info
rxcui = []
rxcui_name = []

# outliers removed
outliers = []

# linear estimate
linear_est = []

# linear estimate stats
linear_params_const = []
linear_params_predictor = []
linear_rsquared = []
linear_pvalues_const = []
linear_pvalues_predictor = []
linear_conf_int_lower = []
linear_conf_int_upper = []

# loop through list of RxCUI in df_top list and run regression analysis on monthly re-sampled data
for i in df_ols_monthly.columns:
    # store RxCUI information
    df_top_rxcui = i
    df_top_rxcui_name = df_top.loc[(df_top['rxcui'] == df_top_rxcui),'drug_name'].unique()[0]
    
    # subset monthly re-sampled data on RxCUI
    forecast = pd.DataFrame(df_ols_monthly[df_top_rxcui])
    forecast = forecast.dropna()
    forecast['predictor'] = range(1, len(forecast)+1)
    forecast.columns = ['cpq','predictor']
    
    # replace infinity with NaN and then remove any NaN's
    forecast = forecast.replace([np.inf, -np.inf], np.nan)
    forecast = forecast.dropna()
    
    # define model inputs/outputs
    y = forecast['cpq'].reset_index(drop=True)  # response
    X = forecast['predictor'].reset_index(drop=True)  # predictor
    X = sm.add_constant(X)  # add a constant term to the predictor

    # fit linear model
    est = sm.OLS(y, X)
    est = est.fit()
    est.summary()
    
    # count statistically significant outliers
    outlier_count = np.abs(np.sum(pd.Series(-(est.outlier_test()['bonf(p)'] < 0.05))-1))

    # if there were outliers then remove them...
    # ... then rerun the outlier test again against filtered dataframe
    if outlier_count > 0:
        # remove outliers from forecast dataframe
        forecast_inliers = pd.Series(-(est.outlier_test()['bonf(p)'] < 0.05))
        forecast = forecast.reset_index().loc[forecast_inliers,].set_index('service_date')
        
        # rerun model
        y = forecast['cpq'].reset_index(drop=True)  # response
        X = forecast['predictor'].reset_index(drop=True)  # predictor
        X = sm.add_constant(X)  # Adds a constant term to the predictor
        est = sm.OLS(y, X)
        est = est.fit()
        est.summary()

        # count statistically significant outliers
        outlier_count2 = np.abs(np.sum(pd.Series(-(est.outlier_test()['bonf(p)'] < 0.05))-1))

        # combine tests results from outlier test
        outlier_count = outlier_count + outlier_count2
        
        if outlier_count2 > 0:
            # remove outliers from forecast dataframe
            forecast_inliers = pd.Series(-(est.outlier_test()['bonf(p)'] < 0.05))
            forecast = forecast.reset_index().loc[forecast_inliers,].set_index('service_date')

            # rerun model
            y = forecast['cpq'].reset_index(drop=True)  # response
            X = forecast['predictor'].reset_index(drop=True)  # predictor
            X = sm.add_constant(X)  # Adds a constant term to the predictor
            est = sm.OLS(y, X)
            est = est.fit()
            est.summary()
            
    # append Rx information
    rxcui.append(i)
    rxcui_name.append(df_top_rxcui_name)
    
    # forecast points
    forecast_points.append(forecast)
    
    # append count of statistically significant outliers to outlier list
    outliers.append(outlier_count)
    
    # write model performance metrics to lists
    linear_params_const.append(est.params[0])
    linear_params_predictor.append(est.params[1])
    linear_rsquared.append(est.rsquared)
    linear_pvalues_const.append(est.pvalues[0])
    linear_pvalues_predictor.append(est.pvalues[1])
    linear_conf_int_lower.append(est.conf_int()[0][0])
    linear_conf_int_upper.append(est.conf_int()[1][0])
    
    # save linear model
    linear_est.append(est)
```

```python
### Elbow curve to assess 'optimal' number of neighborhoods that best describes primary member profiles
# Cluster data into K=1..10 clusters

# K, KM, centroids,D_k,cIdx,dist,avgWithinSS = kmeans.run_kmeans(X,10)

K = range(1,20)

# scipy.cluster.vq.kmeans
KM = [kmeans(gross_cost_cma,k) for k in K] # apply kmeans 1 to 5
centroids = [cent for (cent,var) in KM]   # cluster centroids

D_k = [cdist(gross_cost_cma, cent, 'euclidean') for cent in centroids]

cIdx = [np.argmin(D,axis=1) for D in D_k]
dist = [np.min(D,axis=1) for D in D_k]
avgWithinSS = [sum(d)/gross_cost_cma.shape[0] for d in dist]  

kIdx = 4
# plot elbow curve
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.plot(K, avgWithinSS, 'b*-')
ax.plot(K[kIdx], avgWithinSS[kIdx], marker='o', markersize=12, 
      markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average within-cluster sum of squares')
tt = plt.title('Elbow Plot for k-Means: ' + rx_ndc) 

plt.savefig(directory+'/3_kmeans_elbow_plot.png', bbox_inches='tight')
```
