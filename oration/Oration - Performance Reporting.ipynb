# Campaign Performance Report

```python
import datetime 
import numpy as np
import pandas as pd
from pandas.io import gbq
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

```python
###############################################################################
# Query AdWords.ad from Google BigQuery API

projectid = 'aerial-jigsaw-823'

log_query = """SELECT
log.external_member_id AS external_member_id,
ees.oration_id AS oration_id,
ees.primary_oration_id AS primary_oration_id,
ees.org_name AS org_name,
log.datetime AS datetime,
IF(log.event_name CONTAINS 'member_invited',1,0) AS member_email,
IF(log.event_name CONTAINS 'dependent_invited',1,0) AS dependent_email,
IF(log.event_name CONTAINS 'member_visit',1,0) AS member_visit,
IF(log.event_name CONTAINS 'registration_complete',1,0) AS registration_complete,
IF(log.event_name CONTAINS 'order_submitted',1,0) AS order_submitted,
IF(log.event_name CONTAINS 'campaign_call',1,0) AS campaign_call
FROM
log.detail log LEFT JOIN
(SELECT
primary_oration_id AS primary_oration_id,
oration_id AS oration_id,
member_id AS external_member_id,
org_short_name AS org_name,
date_part AS yob,
cnt AS dependents,
relationship AS relationship
FROM
ees.enrollment
GROUP BY
primary_oration_id, oration_id, external_member_id, org_name, yob, dependents, relationship) ees
ON
log.external_member_id = ees.external_member_id
WHERE 
(event_name CONTAINS 'member_invited' OR
event_name CONTAINS 'dependent_invited' OR
event_name CONTAINS 'member_visit' OR
event_name CONTAINS 'registration_complete' OR
event_name CONTAINS 'order_submitted' OR
event_name CONTAINS 'campaign_call') 
AND
ees.org_name NOT IN ('ORG0006','ORG0007')
GROUP BY
external_member_id,  oration_id, primary_oration_id, org_name, datetime, 
member_email, dependent_email, member_visit, registration_complete, order_submitted, campaign_call;"""

optimization_query = """SELECT
b.oration_id AS oration_id,
b.primary_oration_id AS primary_oration_id,
b.org_short_name AS org_name,
b.date_part AS yob,
b.cnt AS rel_cnt,
b.relationship AS rel,
a.custom_name AS drug_name,
a.strength AS strength,
a.lf_chain AS lf_chain,
a.opt_chain AS opt_chain,
a.opt_quantity AS opt_quantity,
a.opt_days_supply AS opt_days_supply,
a.lf_days_supply AS lf_days_supply,
a.lf_fills_per_year AS lf_fills_per_year,
a.order_status AS order_status,
a.order_date AS order_date,
SUM(lf_member_cost) AS lf_member_cost,
SUM(lf_org_cost) AS lf_org_cost,
SUM(a.member_annual_cost) AS member_annual_cost,
SUM(a.member_annual_savings) AS member_annual_savings,
AVG(a.member_cost_per_fill) AS member_cost_per_fill,
AVG(a.member_savings_per_fill) AS member_savings_per_fill,
SUM(a.org_annual_cost) AS org_annual_cost,
SUM(a.org_annual_savings) AS org_annual_savings, 
AVG(a.org_cost_per_fill) AS org_cost_per_fill,
AVG(a.org_savings_per_fill) AS org_savings_per_fill,
SUM(a.rebate) AS rebate
FROM
opt.summary a LEFT JOIN ees.enrollment b
ON
a.oration_id = b.oration_id
AND 
a.primary_oration_id = b.primary_oration_id
AND
a.org_name = b.org_short_name
WHERE
b.org_short_name NOT IN ('ORG0006','ORG0007')
GROUP BY
oration_id, primary_oration_id, org_name, yob, rel_cnt, rel, drug_name, strength,
lf_chain, opt_chain, opt_quantity, opt_days_supply, lf_days_supply, lf_fills_per_year, order_status, order_date;"""

campaign_query = """SELECT * 
FROM
ees.campaign"""
    
log = pd.read_gbq(log_query, project_id = projectid, reauth=False)
optimization = pd.read_gbq(optimization_query, project_id = projectid, reauth=False)
campaign = pd.read_gbq(campaign_query, project_id = projectid, reauth=False)
```

```python
#####################################################################
# Preprocess Campaign Data
#####################################################################
```

```python
# Drop records where campaign_name is null
campaign = campaign.loc[-campaign['campaign_name'].isnull(),]

# Encode org_name based on campaign name logic (if contains, else NaN)
campaign['org_name'] = 'NaN'
campaign.loc[campaign['campaign_name'].str.contains('YMKT'),'org_name'] = 'YMKT'
campaign.loc[campaign['campaign_name'].str.contains('PTSN'),'org_name'] = 'PTSN'
campaign.loc[campaign['campaign_name'].str.contains('MAF'),'org_name'] = 'MAF'
```

```python
#####################################################################
# Preprocess Optimization Data
#####################################################################
```

```python
# Calculate current cost (last fill cost)
optimization['lf_member_cost'] = optimization['lf_fills_per_year'] * optimization['lf_member_cost']
optimization['lf_org_cost'] = optimization['lf_fills_per_year'] * optimization['lf_org_cost']
optimization['lf_total_cost'] = optimization['lf_member_cost'] + optimization['lf_org_cost']

# Calculate proposed cost (optimized cost)
optimization['o_member_cost'] = optimization['member_annual_cost']
optimization['o_org_cost'] = optimization['org_annual_cost']
optimization['o_total_cost'] = optimization['o_member_cost'] + optimization['o_org_cost']

# Calculate savings (last fill - optimized cost)
optimization['member_annual_savings_calc'] = optimization['lf_member_cost'] - optimization['o_member_cost']
optimization['org_annual_savings_calc'] = optimization['lf_org_cost'] - optimization['o_org_cost']
optimization['total_annual_savings_calc'] = optimization['lf_total_cost'] - optimization['o_total_cost']

# Calculate difference between system reported and manually calculated savings
optimization['total_calc_delta'] = ((optimization['member_annual_savings_calc'] - optimization['member_annual_savings']) + 
                            (optimization['org_annual_savings_calc'] - optimization['org_annual_savings']))
```

```python
# Create 'date' field from 'datetime' field, by stripping time
log['date'] = log['datetime'].apply(lambda x: datetime.date(x.year,x.month,x.day))
log_agg = log.groupby(['primary_oration_id','oration_id','org_name','date']).sum()[[
        'member_email','dependent_email','member_visit','registration_complete','order_submitted','campaign_call']].reset_index()

log_agg.describe()
```

```python
#####################################################################
# Create df for Product Performance Report
#####################################################################
```

```python
# Merge campaign and optimization files
df = pd.merge(optimization[['primary_oration_id','oration_id','org_name',
                            'rel']],
              campaign[['primary_oration_id','oration_id','org_name',
                       'campaign_name','has_email']],
              on=['primary_oration_id','oration_id','org_name'], 
              how='outer').drop_duplicates().sort(columns=['primary_oration_id','oration_id']).reset_index(drop=True)

df = pd.merge(log_agg, df, 
         on=['primary_oration_id','oration_id','org_name'], 
         how='outer').drop_duplicates().sort(columns=['primary_oration_id','oration_id']).reset_index(drop=True)
```

```python
executive_campaign = df.groupby(['org_name','campaign_name','date'])[
'member_email','dependent_email','member_visit',
'registration_complete','order_submitted','campaign_call'].sum().reset_index().pivot_table(index='date',
                                                                   columns=['org_name','campaign_name'], 
                                                                   aggfunc=np.sum, fill_value=0)

executive_campaign.head()
```

```python
dates = []

for i in range(0,len(log_agg['date'])):
    dates.append('Date.UTC(' + pd.DatetimeIndex(log_agg['date']).year.astype(str)[i] + ', ' + pd.DatetimeIndex(log_agg['date']).month.astype(str)[i] + ', ' + pd.DatetimeIndex(log_agg['date']).day.astype(str)[i] + ')')

primary_invites = log_agg['primary_invite'].tolist()
dependent_invites = log_agg['dependent_invite'].tolist()
visits = log_agg['visits'].tolist()
registrations = log_agg['registrations'].tolist()
orders = log_agg['orders'].tolist()

primary_invites = [[i, j] for i, j in zip(dates, primary_invites)]
dependent_invites = [[i, j] for i, j in zip(dates, dependent_invites)]
visits = [[i, j] for i, j in zip(dates, visits)]
registrations = [[i, j] for i, j in zip(dates, registrations)]
orders = [[i, j] for i, j in zip(dates, orders)]
```

```python
opt_summary = opts.groupby(['org_short_name'])[['c_member_cost','c_org_cost','c_total_cost',
                                        'p_member_cost','p_org_cost','p_total_cost',
                                        'member_annual_savings_calc',
                                        'org_annual_savings_calc',
                                        'total_annual_savings_calc']].sum().reset_index()

opt_summary['pct_member_savings'] = opt_summary['member_annual_savings_calc'] / opt_summary['c_member_cost']
opt_summary['pct_org_savings'] = opt_summary['org_annual_savings_calc'] / opt_summary['c_org_cost']
opt_summary['pct_total_savings'] = opt_summary['total_annual_savings_calc'] / opt_summary['c_total_cost']

opt_summary_org = opt_summary['org_short_name'].tolist()
opt_summary_c_total_cost = np.round(opt_summary['c_total_cost'],0).astype(int)
opt_summary_p_total_cost = np.round(opt_summary['p_total_cost'],0).astype(int)
opt_summary_total_savings = np.round(opt_summary['pct_total_savings'],4).astype(float)

opt_summary_c_total_cost = [[i, j] for i, j in zip(opt_summary_org, opt_summary_c_total_cost)]
opt_summary_p_total_cost = [[i, j] for i, j in zip(opt_summary_org, opt_summary_p_total_cost)]
opt_summary_total_savings = [[i, j] for i, j in zip(opt_summary_org, opt_summary_total_savings)]

opt_summary
```

```python
opt_summary_by_rel = opts.groupby(['org_short_name','rel'])[['c_member_cost','c_org_cost','c_total_cost',
                                        'p_member_cost','p_org_cost','p_total_cost',
                                        'member_annual_savings_calc',
                                        'org_annual_savings_calc',
                                        'total_annual_savings_calc']].sum().reset_index()

opt_summary_by_rel['pct_member_savings'] = opt_summary_by_rel['member_annual_savings_calc'] / opt_summary_by_rel['c_member_cost']
opt_summary_by_rel['pct_org_savings'] = opt_summary_by_rel['org_annual_savings_calc'] / opt_summary_by_rel['c_org_cost']
opt_summary_by_rel['pct_total_savings'] = opt_summary_by_rel['total_annual_savings_calc'] / opt_summary_by_rel['c_total_cost']

opt_summary_by_rel_org = opt_summary_by_rel['org_short_name'].tolist()
opt_summary_by_rel_rel = opt_summary_by_rel['rel'].tolist()
opt_summary_by_rel_c_total_cost = np.round(opt_summary_by_rel['c_total_cost'],0).astype(int)
opt_summary_by_rel_p_total_cost = np.round(opt_summary_by_rel['p_total_cost'],0).astype(int)
opt_summary_by_rel_total_savings = np.round(opt_summary_by_rel['pct_total_savings'],4).astype(float)

opt_summary_by_rel_c_total_cost = [[i, j, k] for i, j, k in zip(opt_summary_by_rel_org, opt_summary_by_rel_rel, opt_summary_by_rel_c_total_cost)]
opt_summary_by_rel_p_total_cost = [[i, j, k] for i, j, k in zip(opt_summary_by_rel_org, opt_summary_by_rel_rel, opt_summary_by_rel_p_total_cost)]
opt_summary_by_rel_total_savings = [[i, j, k] for i, j, k in zip(opt_summary_by_rel_org, opt_summary_by_rel_rel, opt_summary_by_rel_total_savings)]

opt_summary_by_rel
```
