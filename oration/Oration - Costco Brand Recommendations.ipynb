```python
import numpy as np
import pandas as pd
from pandas.io import gbq
import matplotlib.pyplot as plt
```

```python
# Load optimization file as 'opts'
projectid = 'aerial-jigsaw-823'
opt_query = """SELECT * FROM opt.summary"""
    
# Optimization file (to identify converted members)
opts = pd.read_gbq(opt_query, project_id = projectid, reauth=False)

# Filter for unique converted Oration IDs
conv = opts.loc[
(opts['order_status'] == 'Completed') |
(opts['order_status'] == 'Active') |
(opts['order_status'] == 'Pending'),'oration_id'].unique()

conv = conv.astype(float).astype(str)
```

```python
# Load active brand scripts as 'df'
df = pd.read_csv('/Users/smacrae/Documents/adhoc/costco/brand_drugs/brand-name-prescriptions-2015-01-15.csv',
                  sep=';', index_col=None, parse_dates=True, infer_datetime_format=True)

# Flag active brand scripts where member has converted
df['convert'] = df['oration_id'].astype(str).isin(conv)
```

```python
print pd.to_datetime(df['service_date']).min()
print pd.to_datetime(df['service_date']).max()
```

```python
# # Distribution of days supply (note the large number of records with small)
# %matplotlib inline
# df.groupby(['days_supply']).apply(len).plot(kind='bar', figsize=(12,6))
# plt.show()
```

```python
# Calculate SUM(gross_cost, quantity) and LEN(scripts)
df_agg = pd.DataFrame(
pd.DataFrame(df.groupby(['convert','drug_name','rxcui','ndc','strength','days_supply'])[['gross_cost', 'quantity']].apply(sum)).join(
pd.DataFrame(df.groupby(['convert','drug_name','rxcui','ndc','strength','days_supply'])[['drug_name']].apply(len)))
).reset_index()

# Update column names
df_agg.columns = ['convert','drug_name','rxcui','ndc','strength','days_supply','total_gross_cost','total_quantity','total_scripts']

# Calculate number of refills per year
df_agg['refills_year'] = np.floor(365/df_agg['days_supply'])

# Calculate estimated annual revenue
df_agg['revenue_365'] = df_agg['refills_year'] * df_agg['total_gross_cost']

# Flag acute vs. chronic scripts based on days supply heuristic
df_agg['chronic'] = df_agg['days_supply'] >= 28

df_agg = df_agg[['drug_name','rxcui','ndc','strength','days_supply','refills_year',
        'total_scripts','total_quantity','total_gross_cost','revenue_365','chronic','convert']]

# Print summary revenue statistics
print df_agg.head()
print 'Est. annual gross revnenue (active scripts):', df_agg.loc[(df_agg['chronic']) == 1,'revenue_365'].sum()
print 'Est. annual gross revnenue (active scripts / converted members):', df_agg.loc[(df_agg['convert']) & (df_agg['chronic']) == 1,'revenue_365'].sum()
```

```python
# Set BIC directory
bic_directory = '/Users/smacrae/Documents/adhoc/pricing_model/data/results/ymkt-rxcui-best_in_class_v2.csv'
 
# Read BIC rates (from YMKT's analysis)
bic = pd.read_csv(bic_directory)[['rxcui','drug_strength','cpq_bic']]
bic.columns = ['rxcui','strength','costper_quantity_bic']

bic
```

```python
pd.DataFrame(bic)
```

```python
# Calculate final BIC as minimum between observed and calculated BIC
bic = []

for i in range(0,len(df_agg)):
    if df_agg.loc[i,'revenue_365'] >= df_agg.loc[i,'revenue_365_bic']:
        bic.append(df_agg.loc[i,'revenue_365_bic'].astype(float))
    else:
        bic.append(df_agg.loc[i,'revenue_365'].astype(float))
        
df_agg['bic'] = pd.Series(bic)

# Reverse engineer gross cost per quantity
df_agg['gross_cost_per_script'] = df_agg['bic'] / df_agg['refills_year'] / df_agg['total_scripts']

# Subset df_agg on converted members with chronic scripts, only
members = df_agg.loc[(df_agg['chronic'] == 1) & (df_agg['convert'] == 1),]

print members['bic'].sum()
```

```python
final = members.groupby(['drug_name','rxcui','ndc','strength','days_supply','gross_cost_per_script'])[['bic','total_scripts']].apply(sum).reset_index().sort(columns='bic', ascending=False).reset_index(drop=True)
final = final[[u'drug_name', u'rxcui', u'ndc', u'strength', u'days_supply',u'total_scripts',u'gross_cost_per_script', u'bic']]

final.columns = [u'drug_name', u'rxcui', u'ndc', u'strength', u'days_supply',u'total_scripts',u'gross_cost_per_script', u'estimated_annual_revenue']
final
```

```python
top_10 = df.loc[(df['drug_name'].isin(final['drug_name'].unique()[0:10])) &
           df['convert']==True,['drug_name','rxcui','ndc','strength','quantity','days_supply','gross_cost']]

top_10 = pd.DataFrame(top_10.groupby(['drug_name','rxcui','ndc','strength','quantity','days_supply'])['gross_cost'].apply(min)).reset_index()

top_10['refills_year'] = np.floor(365/top_10['days_supply'])

top_10['revenue_365'] = top_10['refills_year'] * top_10['gross_cost']

top_10
```
