```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import os
```

```python
# get list of files in log level directory
logpath = r'/Users/smacrae/Documents/data_warehouse/ees'
loglist = os.listdir(logpath)

# # get list of files in optimization directory
# optpath = r'/Users/smacrae/Documents/data_warehouse/opt'
# optlist = os.listdir(optpath)
```

```python
# create list of strings for the directory path of each files in Adgooroo directory
logdata = []
for dir_entry in os.listdir(logpath):
    dir_entry_path = os.path.join(logpath, dir_entry)
    if os.path.isfile(dir_entry_path):
        logdata.append(dir_entry_path)
```

```python
# write each log file into a list of pandas dataframes
pieces = []

for i in range(0, len(logdata)):
    log = pd.read_csv(logdata[i], sep=',', index_col=None, parse_dates=True, infer_datetime_format=True)    
    log['directory'] = logdata[i]
    pieces.append(log)
```

```python
# combine list of dataframes into a single dataframe
logs = pd.concat(pieces, ignore_index=True)
```
