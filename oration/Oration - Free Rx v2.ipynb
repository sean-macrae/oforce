```python
# Combine 1) claims file, 2) costco prices, and 3) rxnorm datasets by ndc
df = pd.merge(df, prices, on=['ndc'], how='left', suffixes=['_df','_prices'])
df = pd.merge(df, rx, on=['ndc'], how='left', suffixes=['_df','_rx'])
df = df.drop_duplicates()
```

```python
# Last overall fill date
df['service_dt'] = pd.to_datetime(df['service_dt'])
df['last_fill_date'] = df['service_dt'].max()
df['service_dt'].max()
```

```python
# Filter days supply >20 and <365
df = df.loc[(df['days_supply'] > 20) & (df['days_supply'] < 365),].reset_index(drop=True)

# Filter quantity > 0
df = df.loc[df['quantity_df'] > 0,].reset_index(drop=True)

# Filter for Capsules & Tablets only
# df = df.loc[(df['dose_form_df'] == 'Oral Capsule') | (df['dose_form_df'] == 'Oral Tablet'), ].reset_index(drop=True)
```

```python
# 1) Calculate "active window"
df['active_window'] = 0

df.loc[df['days_supply'] <= 31,'active_window'] = pd.Timedelta('90 days')
df.loc[(df['days_supply'] > 31) & (df['days_supply'] <= 91),'active_window'] = pd.Timedelta('180 days')
df.loc[df['days_supply'] > 91,'active_window'] = pd.Timedelta('360 days')

df['active_date'] = pd.to_datetime(df['last_fill_date']) - df['active_window']

# Service date >= Active date
df['rule_1'] = pd.Series(pd.to_datetime(df['active_date']) <= pd.to_datetime(df['service_dt']))
```

```python
# 2) Calculate "next service date window"
df_last_fill = df[['primary_oration_id','oration_id','rxcui_rx',
                   'service_dt']].groupby([
                   'primary_oration_id','oration_id','rxcui_rx']).max().reset_index(drop=False)

df_last_fill.columns = ['primary_oration_id','oration_id','rxcui_rx','service_dt_max']

df = pd.merge(df, df_last_fill, on=['primary_oration_id','oration_id','rxcui_rx'])
```

```python
# Calculate next fill date
fill_window = []

for i in range(0, len(df)):
    fill_window.append(pd.Timedelta(str(df['days_supply'][i])+' days'))

df['fill_window'] = fill_window

for i in range(0, len(df)):
    df['fill_date'] = pd.to_datetime(df['service_dt_max'][i]) + df['fill_window']

df['fill_date'] = pd.to_datetime(df['service_dt']) + df['fill_window']

df['rule_2'] = pd.Series(df['fill_date'] >= df['last_fill_date'] - pd.Timedelta(str('45 days')))
```

```python
# Filter for active drugs
df = df.loc[(df['rule_1'] == True) & (df['rule_2'] == True), ]
```

```python
final = whales_updated[['primary_oration_id','oration_id','service_dt_max',
                        'custom_name','o_quantity','days_supply','ndc','rxcui_rx','is_brand',
                'price_x']].groupby(['primary_oration_id','oration_id','service_dt_max',
                        'custom_name','o_quantity','days_supply','ndc','rxcui_rx','is_brand']).max().reset_index()

patient_pay = []

import math

for i in final.index:
    if math.isnan(np.float64(final['price_x'][i])):
        patient_pay.append(np.nan)
    elif ((final.loc[i,'is_brand']==True) & (np.float64(final.loc[i,'price_x'])*0.20 >= 50.00)):
        patient_pay.append(50.00)
    elif ((final.loc[i,'is_brand']==True) & (np.float64(final.loc[i,'price_x'])*0.20 < 50.00)):
        patient_pay.append(np.float64(final.loc[i,'price_x'])*0.20)
    elif ((final.loc[i,'is_brand']==False) & (np.float64(final.loc[i,'price_x']) >= 20.00)):
        patient_pay.append(20.00)
    elif ((final.loc[i,'is_brand']==False) & (np.float64(final.loc[i,'price_x']) < 20.00)):
        patient_pay.append(np.float64(final.loc[i,'price_x']))

final['patient_pay'] = patient_pay

final.loc[final['patient_pay'] > 100.00,'patient_pay'] = 100.00

final['org_pay'] = np.float64(final['price_x']) - final['patient_pay']
```

```python
# De-dupe final

final = pd.merge(
final[['primary_oration_id','oration_id',
      'custom_name','service_dt_max']].groupby(['primary_oration_id','oration_id','custom_name']).max().reset_index(),
final,
on=['primary_oration_id','oration_id',
      'custom_name','service_dt_max'],
how='left'
)

final = final.drop_duplicates()
```

```python
hipo = pd.merge(final, whales_updated, 
         on=['primary_oration_id','oration_id',
             'custom_name','rxcui_rx','o_quantity'],
         how='left')[['primary_oration_id','oration_id','rxcui_rx','custom_name','is_brand_x',
                                        'o_quantity','days_supply_x','price_x_x','patient_pay','org_pay',
                                        'quantity_df','total_amount','patient_pay_calculated','net_due_amount']]

hipo.columns = ['primary_oration_id','oration_id','rxcui','drug_name','is_brand',
                                        'costco_quantity','days_supply','costco_price','costco_patient_pay','costco_org_pay',
                                        'lf_quantity','lf_price','lf_patient_pay','lf_org_pay']

hipo['costco_days_supply'] = hipo['costco_quantity'] / (hipo['lf_quantity'] / hipo['days_supply'])
hipo['lf_refills'] = np.floor(365/hipo['days_supply'])
hipo['costco_refills'] = np.floor(365/hipo['costco_days_supply'])

hipo['o_annual_cost_org'] = np.round(hipo['costco_org_pay']*hipo['costco_refills'],2)
hipo['o_annual_cost_member'] = np.round(hipo['costco_patient_pay']*hipo['costco_refills'],2)
hipo['o_annual_cost'] = hipo['o_annual_cost_org'] + hipo['o_annual_cost_member']

hipo['annual_savings_org'] = np.round((hipo['lf_org_pay']*hipo['lf_refills']) - (hipo['costco_org_pay']*hipo['costco_refills']),2)
hipo['annual_savings_member'] = np.round((hipo['lf_patient_pay']*hipo['lf_refills']) - (hipo['costco_patient_pay']*hipo['costco_refills']),2)
hipo['annual_savings'] = np.round(hipo['annual_savings_org'] + hipo['annual_savings_member'],2)
```

```python
# Add count of the number of times a drug appears by member (for deduping purposes)
hipo_dupes = pd.DataFrame(hipo[['primary_oration_id','oration_id',
                   'drug_name']].groupby(['primary_oration_id','oration_id','drug_name']).apply(len)).reset_index()
hipo_dupes.columns = ['primary_oration_id','oration_id','drug_name','count']
hipo = pd.merge(hipo, hipo_dupes, on=['primary_oration_id','oration_id','drug_name'], how='left')

# HiPo drop duplicates
hipo = hipo.drop_duplicates()
```
