```python
import pandas as pd
import numpy as np
import random

# metrics

# gross_cost
mu, sigma = 140, 180
gross_cost = []
for i in range(0,1000):
    gross_cost.append(np.round(random.gauss(mu, sigma),2))
gross_cost = pd.Series(gross_cost)
    
# company cost
mu, sigma = 100, 120
company_cost = []
for i in range(0,1000):
    company_cost.append(np.round(random.gauss(mu, sigma),2))
company_cost = pd.Series(company_cost)
    
# enrollee cost
mu, sigma = 40, 60
enrollee_cost = []
for i in range(0,1000):
    enrollee_cost.append(np.round(random.gauss(mu, sigma),2))
enrollee_cost = pd.Series(enrollee_cost)

# savings
mu, sigma = 20, 80
savings = []
for i in range(0,1000):
    savings.append(np.round(random.gauss(mu, sigma),2))
savings = pd.Series(savings)

# company savings
mu, sigma = 0, 50
company_savings = []
for i in range(0,1000):
    company_savings.append(np.round(random.gauss(mu, sigma),2))
company_savings = pd.Series(company_savings)

# enrollee savings
mu, sigma = 0, 30
enrollee_savings = []
for i in range(0,1000):
    enrollee_savings.append(np.round(random.gauss(mu, sigma),2))
enrollee_savings = pd.Series(enrollee_savings)

# enrollees
enrollees = [1] * 1000
enrollees = pd.Series(enrollees)

# dependents
mu, sigma = 0, 3
dependents = []
for i in range(0,1000):
    dependents.append(np.floor(random.gauss(mu, sigma)))
dependents = pd.Series(dependents)
dependents[dependents < 0] = 0
```

```python
# dimensions

# company
def random_ints(num, lower=0, upper=5):
    return [random.randrange(lower,upper+1) for i in range(num)]
company = pd.Series(random_ints(1000))

# industry
def random_ints(num, lower=0, upper=15):
    return [random.randrange(lower,upper+1) for i in range(num)]
industry = pd.Series(random_ints(1000))

# gender
def random_ints(num, lower=0, upper=1):
    return [random.randrange(lower,upper+1) for i in range(num)]
gender = pd.Series(random_ints(1000))

# pharmacy
def random_ints(num, lower=0, upper=100):
    return [random.randrange(lower,upper+1) for i in range(num)]
pharmacy = pd.Series(random_ints(1000))

# chain
def random_ints(num, lower=0, upper=25):
    return [random.randrange(lower,upper+1) for i in range(num)]
chain = pd.Series(random_ints(1000))

# pbm
def random_ints(num, lower=0, upper=13):
    return [random.randrange(lower,upper+1) for i in range(num)]
pbm = pd.Series(random_ints(1000))

# state
def random_ints(num, lower=0, upper=51):
    return [random.randrange(lower,upper+1) for i in range(num)]
state = pd.Series(random_ints(1000))

# dma
def random_ints(num, lower=0, upper=40):
    return [random.randrange(lower,upper+1) for i in range(num)]
dma = pd.Series(random_ints(1000))

# county
def random_ints(num, lower=0, upper=500):
    return [random.randrange(lower,upper+1) for i in range(num)]
county = pd.Series(random_ints(1000))

# zip
def random_ints(num, lower=0, upper=1000):
    return [random.randrange(lower,upper+1) for i in range(num)]
zip = pd.Series(random_ints(1000))
```

```python
df = pd.concat([gross_cost, company_cost],axis=1)
df = pd.concat([df, enrollee_cost],axis=1)
df = pd.concat([df, savings],axis=1)
df = pd.concat([df, company_savings],axis=1)
df = pd.concat([df, enrollee_savings],axis=1)
df = pd.concat([df, enrollees],axis=1)
df = pd.concat([df, dependents],axis=1)
df = pd.concat([df, company],axis=1)
df = pd.concat([df, industry],axis=1)
df = pd.concat([df, gender],axis=1)
df = pd.concat([df, pharmacy],axis=1)
df = pd.concat([df, chain],axis=1)
df = pd.concat([df, pbm],axis=1)
df = pd.concat([df, state],axis=1)
df = pd.concat([df, dma],axis=1)
df = pd.concat([df, county],axis=1)
df = pd.concat([df, zip],axis=1)

df.columns = ['gross_cost','company_cost','enrollee_cost',
            'savings','company_savings','enrollee_savings',
            'enrollees','dependents',
            'company','industry','gender','pharmacy',
            'chain','pbm','state','dma','county','zip']

df.to_csv(r'/Users/smacrae/Desktop/dummy_data.csv', index=False)
```
