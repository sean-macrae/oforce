```python
# Elbow curve to assess 'optimal' number of neighborhoods that best describes primary member profiles

##### cluster data into K=1..10 clusters #####
#K, KM, centroids,D_k,cIdx,dist,avgWithinSS = kmeans.run_kmeans(X,10)

K = range(1,12)

# scipy.cluster.vq.kmeans
KM = [kmeans(signals,k) for k in K] # apply kmeans 1 to 10
centroids = [cent for (cent,var) in KM]   # cluster centroids

D_k = [cdist(signals, cent, 'euclidean') for cent in centroids]

cIdx = [np.argmin(D,axis=1) for D in D_k]
dist = [np.min(D,axis=1) for D in D_k]
avgWithinSS = [sum(d)/signals.shape[0] for d in dist]  

kIdx = 7
# plot elbow curve
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(K, avgWithinSS, 'b*-')
ax.plot(K[kIdx], avgWithinSS[kIdx], marker='o', markersize=12, 
      markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average within-cluster sum of squares')
tt = plt.title('Elbow for K-Means clustering') 
plt.show()
```

```python
# Dimension Reduction through Principal Component Analysis & Linear Discriminant Analysis

X = signals
y = np.array(pd.get_dummies(df[['order_status']])['order_status_Completed'].tolist())
target_names = np.array(['yes','no'])
```

```python
# PCA
# Principal Component Analysis (PCA) applied to this data identifies 
# the combination of attributes (principal components, or directions in the feature space) 
# that account for the most variance in the data
from sklearn.decomposition import PCA

pca = PCA(n_components=len(X[0]))
X_r = pca.fit(X).transform(X)

print(pca.fit(X).explained_variance_ratio_) 

plt.figure()
for c, i, target_name in zip("rgb", [0, 1], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], c=c, label=target_name)
plt.legend()
plt.title('PCA of Clustering Signals by Conversion Type')
plt.show()
```

```python
# Plot variance explained of PCA decomposition by number of principal components
plt.bar(np.arange(len(X[0])) + .5, pca.fit(X).explained_variance_ratio_.cumsum())
plt.ylim((0, 1))
plt.xlabel('No. of principal components')
plt.ylabel('Cumulative variance explained')
plt.grid(axis = 'y', ls = '-', lw = 1, color = 'white')
plt.show()
```

```python
# PCA Factor loadings
# We can also check out the loadings of the principal component across the signals. 
# What this shows us is how a change in the components relate to the original data. 
# For example a a component going up might cause half the signals in the data to go up and half to go down 
# (it would positively load on some and negatively load on others.) 
# We would expect, intuitively, a factor representing “beahvior,” as we think our first component does, 
# to load on our stocks in the same direction, and roughly the same magnitude. And this is basically what we see.

# Plot factor loadings of 1st component
plt.bar(np.arange(len(X[0])), pca.fit(X).components_[0])
plt.show()
```

```python
# LDA
# Linear Discriminant Analysis (LDA) tries to identify 
# attributes that account for the most variance between classes. 
# In particular, LDA, in contrast to PCA, is a supervised method, using known class labels.
from sklearn.lda import LDA

lda = LDA(n_components=len(X[0]))
X_r2 = lda.fit(X, y).transform(X)

plt.figure()
for c, i, target_name in zip("rgb", [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], c=c, label=target_name)
plt.legend()
plt.title('LDA of Clustering Signals by Conversion Type')
plt.show()
```
