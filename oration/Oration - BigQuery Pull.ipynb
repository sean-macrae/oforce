```python
# combine logs and employee tables
graph = pd.merge(logs, eess, left_on = 'external_member_id', right_on = 'member_id', left_index=False, right_index=False, how='left')
graph.describe()
```

```python
# filter out bots
graph = graph.loc[((graph['org_short_name'] == 'YMKT') |
(graph['org_short_name'] == 'MAF') |
(graph['org_short_name'] == 'PTSN')),]

graph = graph.reset_index(drop=True)
```

```python
# flag redirect events (where id and timestamp are equal)
graph['flag'] = 0

for i in range(1, len(graph)):
    if ((graph.loc[i,'external_member_id'] == graph.loc[i-1,'external_member_id']) & 
    (graph.loc[i,'datetime'] == graph.loc[i-1,'datetime'])):
        graph.loc[i,'flag'] = 1
        graph.loc[i-1,'flag'] = 1
    else:
        0
        
# remove redirect events (where id and timestamp are equal)
```

```python
# add customer index identifying order of events taken
graph.sort(columns=['external_member_id','datetime'], ascending=True)

index = [1]

for i in range(1, len(graph)):
    if graph.loc[i,'external_member_id'] == graph.loc[i-1,'external_member_id']:
        ix = index[len(index)-1]+1
    else:
        ix = 1
    index.append(ix)
    
graph['index'] = index
```

```python
# calculate difference between events
graph['datetime_diff'] = graph['datetime'].diff()

# error out timestamp differences where index = 1
graph.loc[graph['index']==1,'datetime_diff'] = np.nan

# convert timestamp difference into integer value of seconds
graph['datetime_diff']
```

```python
import datetime

datetime_diff_sec = []

for i in range(0, len(graph['datetime_diff'])):
    if pd.isnull(graph['datetime_diff'][i]):
        datetime_diff_sec.append(np.nan)
    else:
        datetime_diff_sec.append(graph['datetime_diff'][i].total_seconds())
    
graph['datetime_diff_sec'] = datetime_diff_sec
```

```python
import graphlab as gl
from graphlab import SGraph, Vertex, Edge

gl_sframe = gl.SFrame(graph[['external_member_id','event_name','index']])
```

```python
gl_sframe
gl_sframe = gl_sframe.dropna()
```

```python
gl_sframe.show()
```

```python
events = gl_sframe['event_name'].unique()
members = gl_sframe['external_member_id'].unique()
```

```python
g = gl.SGraph()
g = g.add_edges(gl_sframe, src_field='external_member_id', dst_field='event_name')

print "Movie graph summary:\n", g.summary(), "\n"
```

```python
print "Member vertex sample:"
g.get_vertices(ids=members).tail(5)
```

```python
print "Events vertex sample:"
g.get_vertices(ids=events).head(5)
```

```python
selection = ['dlg_missing_rx_open','dlg_norxselected_cancel']

subgraph = gl.SGraph()
subgraph = subgraph.add_edges(g.get_edges(dst_ids=selection),
                              src_field='__src_id', dst_field='__dst_id')
subgraph.show(highlight=selection)
```
