```python
import shapefile
import pandas as pd
```

```python
##################################################################################################################################################################################################
### Read shapefiles
appeals_sf = shapefile.Reader("/Users/Sean/Documents/data_science/blog/san_francisco_neighborhoods/data_sets/property_planning/appeals/Appeals.shp")
historic_sf = shapefile.Reader("/Users/Sean/Documents/data_science/blog/san_francisco_neighborhoods/data_sets/property_planning/art_10_hist_dist/Art10HistDist.shp")
conserve_sf = shapefile.Reader("/Users/Sean/Documents/data_science/blog/san_francisco_neighborhoods/data_sets/property_planning/art_11_conserv_dist/Art11ConservDist.shp")
citylots_sf = shapefile.Reader("/Users/Sean/Documents/data_science/blog/san_francisco_neighborhoods/data_sets/property_planning/citylots/citylots.shp")
addresses_sf = shapefile.Reader("/Users/Sean/Documents/data_science/blog/san_francisco_neighborhoods/data_sets/property_planning/eas_addresses_with_units/eas_addresses_with_units.shp")
heights_sf = shapefile.Reader("/Users/Sean/Documents/data_science/blog/san_francisco_neighborhoods/data_sets/property_planning/heights/Height_and_Bulk_Districts.shp")
permits_sf = shapefile.Reader("/Users/Sean/Documents/data_science/blog/san_francisco_neighborhoods/data_sets/property_planning/misc_permits/MiscPermits.shp")
neighborhoods_sf = shapefile.Reader("/Users/Sean/Documents/data_science/blog/san_francisco_neighborhoods/data_sets/property_planning/planning_neighborhoods/planning_neighborhoods.shp")

### Read shapes in shapefiles
appeals_shapes = appeals_sf.shapes()
historic_shapes = historic_sf.shapes()
conserve_shapes = conserve_sf.shapes()
citylots_shapes = citylots_sf.shapes()
addresses_shapes = addresses_sf.shapes()
heights_shapes = heights_sf.shapes()
permits_shapes = permits_sf.shapes()
neighborhoods_shapes = neighborhoods_sf.shapes()

### Read the bounding box from the 1st shape
appeals_shapes[0].bbox
historic_shapes[0].bbox
conserve_shapes[0].bbox
citylots_shapes[0].bbox
# addresses_shapes[0].bbox
heights_shapes[0].bbox
permits_shapes[0].bbox
neighborhoods_shapes[0].bbox
```

```python
#############################################################################################
### Reading Database Attributes

###################################################################
### Appeals columns
appeals_names = appeals_sf.fields
appeals_columns = []

for i in range(0, len(appeals_names)):
    appeals_columns.append(appeals_names[i][0])

col_len = len(appeals_columns)
appeals_columns = appeals_columns[1:col_len]

###################################################################
### Historic columns
historic_names = historic_sf.fields
historic_columns = []

for i in range(0, len(historic_names)):
    historic_columns.append(historic_names[i][0])
    
col_len = len(historic_columns)
historic_columns = historic_columns[1:col_len]

###################################################################
### Conservation columns
conserve_names = conserve_sf.fields
conserve_columns = []

for i in range(0, len(conserve_names)):
    conserve_columns.append(conserve_names[i][0])
    
col_len = len(conserve_columns)
conserve_columns = conserve_columns[1:col_len]

###################################################################
### Citylots columns
citylots_names = citylots_sf.fields
citylots_columns = []

for i in range(0, len(citylots_names)):
    citylots_columns.append(citylots_names[i][0])
    
col_len = len(citylots_columns)
citylots_columns = citylots_columns[1:col_len]

###################################################################
### Addresses columns
addresses_names = addresses_sf.fields
addresses_columns = []

for i in range(0, len(addresses_names)):
    addresses_columns.append(addresses_names[i][0])
    
col_len = len(addresses_columns)
addresses_columns = addresses_columns[1:col_len]

###################################################################
### Heights columns
heights_names = heights_sf.fields
heights_columns = []

for i in range(0, len(heights_names)):
    heights_columns.append(heights_names[i][0])
    
col_len = len(heights_columns)
heights_columns = heights_columns[1:col_len]

###################################################################
### Permits columns
permits_names = permits_sf.fields
permits_columns = []

for i in range(0, len(permits_names)):
    permits_columns.append(permits_names[i][0])
    
col_len = len(permits_columns)
permits_columns = permits_columns[1:col_len]

###################################################################
### Neighborhoods columns
neighborhoods_names = neighborhoods_sf.fields
neighborhoods_columns = []

for i in range(0, len(neighborhoods_names)):
    neighborhoods_columns.append(neighborhoods_names[i][0])
    
col_len = len(neighborhoods_columns)
neighborhoods_columns = neighborhoods_columns[1:col_len]
```

```python
#############################################################################################
### Read the database records

appeals = pd.DataFrame(appeals_sf.records())
appeals.columns = appeals_columns

historic = pd.DataFrame(historic_sf.records())
historic.columns = historic_columns

conserve = pd.DataFrame(conserve_sf.records())
conserve.columns = conserve_columns

citylots = pd.DataFrame(citylots_sf.records())
citylots.columns = citylots_columns

addresses = pd.DataFrame(addresses_sf.records())
addresses.columns = addresses_columns

heights = pd.DataFrame(heights_sf.records())
heights.columns = heights_columns

permits = pd.DataFrame(permits_sf.records())
permits.columns = permits_columns

neighborhoods = pd.DataFrame(neighborhoods_sf.records())
neighborhoods.columns = neighborhoods_columns
```

```python
#############################################################################################
### Extract shape coordinates into lists

##########################
### Appeals shapes
appeals_shapes_df = []

for i in range(0, len(appeals_shapes)):
    appeals_shapes_df.append(appeals_shapes[i].bbox)

##########################
### Historic shapes
historic_shapes_df = []

for i in range(0, len(historic_shapes)):
    historic_shapes_df.append(historic_shapes[i].bbox)

##########################
### Conservation shapes
conserve_shapes_df = []

for i in range(0, len(conserve_shapes)):
    conserve_shapes_df.append(conserve_shapes[i].bbox)

##########################
### Citylots shapes
# citylots_shapes_df = []

# for i in range(0, len(citylots_shapes)):
#     citylots_shapes_df.append(citylots_shapes[i].bbox)
    
##########################
### Heights shapes
heights_shapes_df = []

for i in range(0, len(heights_shapes)):
    heights_shapes_df.append(heights_shapes[i].bbox)
    
##########################
### Permits shapes
permits_shapes_df = []

for i in range(0, len(permits_shapes)):
    permits_shapes_df.append(permits_shapes[i].bbox)

##########################
### Permits shapes
neighborhoods_shapes_df = []

for i in range(0, len(neighborhoods_shapes)):
    neighborhoods_shapes_df.append(neighborhoods_shapes[i].bbox)
```

```python
#############################################################################################
### Create dataframe objects from lists of shapes

appeals_shapes_df = pd.DataFrame(appeals_shapes_df)
appeals_shapes_df.columns = ['SHAPE_01','SHAPE_02','SHAPE_03','SHAPE_04']

historic_shapes_df = pd.DataFrame(historic_shapes_df)
historic_shapes_df.columns = ['SHAPE_01','SHAPE_02','SHAPE_03','SHAPE_04']

conserve_shapes_df = pd.DataFrame(conserve_shapes_df)
conserve_shapes_df.columns = ['SHAPE_01','SHAPE_02','SHAPE_03','SHAPE_04']

# pd.DataFrame(citylots_shapes_df)

heights_shapes_df = pd.DataFrame(heights_shapes_df)
heights_shapes_df.columns = ['SHAPE_01','SHAPE_02','SHAPE_03','SHAPE_04']

permits_shapes_df = pd.DataFrame(permits_shapes_df)
permits_shapes_df.columns = ['SHAPE_01','SHAPE_02','SHAPE_03','SHAPE_04']

neighborhoods_shapes_df = pd.DataFrame(neighborhoods_shapes_df)
neighborhoods_shapes_df.columns = ['SHAPE_01','SHAPE_02','SHAPE_03','SHAPE_04']
```

```python
#############################################################################################
### Combine shape coordinates into

appeals = pd.concat([appeals, appeals_shapes_df], axis=1)
historic = pd.concat([historic, historic_shapes_df], axis=1)
conserve = pd.concat([conserve, historic_shapes_df], axis=1)
heights = pd.concat([heights, historic_shapes_df], axis=1)
permits = pd.concat([permits, historic_shapes_df], axis=1)
neighborhoods = pd.concat([neighborhoods, historic_shapes_df], axis=1)
```
