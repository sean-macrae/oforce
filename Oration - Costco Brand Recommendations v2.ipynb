```python
import numpy as np
import pandas as pd
from pandas.io import gbq
import matplotlib.pyplot as plt
```

```python
# Load optimization file as 'opts'
projectid = 'aerial-jigsaw-823'
opt_query = """SELECT * FROM opt.summary"""
    
# Optimization file (to identify converted members)
opts = pd.read_gbq(opt_query, project_id = projectid, reauth=False)

# Filter for unique converted Oration IDs
conv = opts.loc[
(opts['order_status'] == 'Completed') |
(opts['order_status'] == 'Active') |
(opts['order_status'] == 'Pending'),'oration_id'].unique()

conv = conv.astype(float).astype(str)
```

```python
opts['org_name'].unique()
```

```python
# Load active brand scripts as 'df'
df = pd.read_csv('/Users/smacrae/Documents/adhoc/costco/brand_drugs/old/brand-name-prescriptions-2015-03-12.csv',
                  sep=',', index_col=None, parse_dates=True, infer_datetime_format=True)

# Calculate CPQ (cost per quantity)
df['cpq'] = df['gross_cost'] / df['quantity']

# Flag active brand scripts where member has converted
df['convert'] = df['oration_id'].astype(str).isin(conv)

# Flag acute vs. chronic scripts based on days supply heuristic
df['chronic'] = df['days_supply'] >= 28
```

```python
# Filter for recent scripts
# df = df.loc[pd.to_datetime(df['service_date']) >= pd.to_datetime('2015-01-01')]
```

```python
# Calculate SUM(gross_cost, quantity) and LEN(scripts)
df_agg = pd.DataFrame(
pd.DataFrame(df.groupby(['convert','chronic','drug_name','rxcui','days_supply'])[['gross_cost', 'quantity']].apply(sum)).join(
pd.DataFrame(df.groupby(['convert','chronic','drug_name','rxcui','days_supply'])[['drug_name']].apply(len)))
).reset_index()

# Update column names
df_agg.columns = ['convert','chronic','drug_name','rxcui','days_supply','total_gross_cost','total_quantity','total_scripts']

# Calculate CPQ (cost per quantity)
df_agg['cpq'] = df_agg['total_gross_cost'].astype(np.float64) / df_agg['total_quantity']

# Calculate number of refills per year
df_agg['refills_year'] = np.floor(365/df_agg['days_supply'])

# Calculate estimated annual revenue
df_agg['revenue_365'] = df_agg['refills_year'] * df_agg['total_gross_cost']

df_agg = df_agg[['drug_name','rxcui','days_supply','refills_year',
        'total_scripts','total_quantity','total_gross_cost','revenue_365','chronic','convert']]

# Print summary revenue statistics
print df_agg.head()
print 'Est. annual gross revnenue (active scripts):', df_agg.loc[(df_agg['chronic']) == 1,'revenue_365'].sum()
print 'Est. annual gross revnenue (active scripts / converted members):', df_agg.loc[(df_agg['convert']) & (df_agg['chronic']) == 1,'revenue_365'].sum()
```

```python
brand_scripts = df_agg.loc[(df_agg['convert']) & (df_agg['chronic']) == 1,
                           ['drug_name','rxcui','total_scripts','total_quantity','days_supply',
                            'refills_year','total_gross_cost','revenue_365']].drop_duplicates()

brand_scripts = brand_scripts.sort(columns='revenue_365', ascending=False).reset_index(drop=True)
brand_scripts['cpq'] = np.round(brand_scripts['total_gross_cost'] / brand_scripts['total_quantity'],2)

brand_top = brand_scripts[0:11]
brand_top
```

```python
brand_raw = df.loc[df['rxcui'].isin(brand_top_rxcui),
           ['drug_name','rxcui','quantity','gross_cost','cpq','service_date']].drop_duplicates().sort(
columns=['rxcui','quantity']).dropna().reset_index(drop=True)
```

```python
def f(x):
    return pd.Series([len(x.index),
                      x.min(),
                      x.quantile(0.20),
                      x.quantile(0.40),
                      x.mean(),
                      x.median(),
                      x.quantile(0.70),
                      x.quantile(0.90),
                      x.max(),
                      x.std()], 
                      index=['Records','Min','20th','40th','Mean','Median','70th','90th','Max','StdDev'])
```

```python
str(100*(1/400.0))+'%'
```

```python
brand_groups = brand_raw.groupby(['drug_name','rxcui'])
brand_cpq_summary = brand_groups['cpq'].apply(f).unstack().reset_index().sort(columns=['rxcui']).reset_index(drop=True)

brand_cpq_summary['Volatility'] = np.round((brand_cpq_summary['StdDev'] / brand_cpq_summary['Median']) * 100,1)
brand_cpq_summary['Skew'] = np.round(brand_cpq_summary['Mean'] - brand_cpq_summary['Median'],1)
brand_cpq_summary.columns = ['drug_name','rxcui','Records',
                             'Min_cpq','20th_cpq','40th_cpq',
                             'Mean_cpq','Median_cpq',
                             '70th_cpq','90th_cpq',
                             'Max_cpq','StdDev_cpq',
                             'Volatility_cpq','Skew_cpq']
brand_cpq_summary
```

```python
brand_cpq_summary = pd.merge(brand_cpq_summary,
pd.DataFrame(brand_groups['service_date'].max()).reset_index(), on=['drug_name','rxcui'], how='left')
```
