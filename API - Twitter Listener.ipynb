```python
# A function for computing lexical diversity
def lexical_diversity(tokens):
    return 1.0*len(set(tokens))/len(tokens) 

# A function for computing the average number of words per tweet
def average_words(statuses):
    total_words = sum([ len(s.split()) for s in statuses ]) 
    return 1.0*total_words/len(statuses)

print lexical_diversity(words)
print lexical_diversity(screen_names)
print lexical_diversity(hashtags)
print average_words(status_texts)
```

```python
# natural language processing toolkit
import re, math, collections, itertools
import nltk, nltk.classify.util, nltk.metrics
from nltk.classify import NaiveBayesClassifier
from nltk.metrics import BigramAssocMeasures
from nltk.probability import FreqDist, ConditionalFreqDist
```

```python
# generate set of tokens (unigrams, bigrams, trigrams) from 'text' field
unigrams = []
bigrams = []
trigrams = []

for i in range(len(formatted_tweets['text'])):
    unigrams.append(nltk.word_tokenize(formatted_tweets['text'][i]))
    
for i in range(len(unigrams)):
    bigrams.append(nltk.bigrams(unigrams[i]))
    
for i in range(len(unigrams)):
    trigrams.append(nltk.trigrams(unigrams[i]))
```

```python
# flatten tokens from list of list, into single list
# create set of unique tokens
# calculate lexical diversity of tokens
unigrams_flattened = [j for i in unigrams for j in i]
unigrams_distinct = set(unigrams_flattened)
print len(unigrams_flattened)
print len(unigrams_distinct)
print lexical_diversity (unigrams_flattened)

bigrams_flattened = [j for i in bigrams for j in i]
bigrams_distinct = set(bigrams_flattened)
print len(bigrams_flattened)
print len(bigrams_distinct)
print lexical_diversity (bigrams_flattened)

trigrams_flattened = [j for i in trigrams for j in i]
trigrams_distinct = set(trigrams_flattened)
print len(trigrams_flattened)
print len(trigrams_distinct)
print lexical_diversity (trigrams_flattened)
```

```python
# import word affinity corpus (part of unigram corpus)
affin = [line.strip('\n') for line in open('/Users/Sean/Documents/data_science/python/sentiment_analysis/AFINN-111.txt')]
```

```python
# create word affinity corpus dictionary

corpus = []

for i in range(len(affin)):
    corpus.append(affin[i].split('\t'))

corpus_dict = dict(corpus)
```

```python
corpus_dict.keys()
```

```python
# create list of 1-grams

tweet_1grams = []

for i in range(len(formatted_tweets)):
    tweet_1grams.append(formatted_tweets['text'][i].split())
```

```python
word_scores = []

i = 0
while i < len(tweet_1grams):
    word_scores.append([])
    i = i + 1

for i in range(len(tweet_1grams)):
    for j in range(len(tweet_1grams[i])):
        try:
            word_scores[i].append(str(corpus_dict.get(str(tweet_1grams[i][j]).lower()).replace('None','0')))
        except:
            pass
```

```python
tweet_scores = []

i = 0
while i < len(word_scores):
    tweet_scores.append([])
    i = i + 1

for i in range(len(word_scores)):
    for j in range(len(word_scores[i])):
        tweet_scores[i].append(int(word_scores[i][j]))
```

```python
scores = []

i = 0
while i < len(tweet_scores):
    scores.append([])
    i = i + 1
    
for i in range(len(tweet_scores)):
    scores[i].append(sum(tweet_scores[i]))
```
