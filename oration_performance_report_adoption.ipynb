```python
# Convert segment timestamp to UNIX timestamp
segment['timestamp'] = pd.to_datetime(segment['timestamp']).astype(np.int64) // 10**9

# Convert customer timestamp to integer
customer['timestamp'] = customer['timestamp'].astype(np.int64)
```

```python
# Add dummies to Customer.io
customer_dummies = pd.get_dummies(customer[['data_variables_eventName',
                                            'eventType']], 
                                  prefix='dummy', prefix_sep='_')

customer_io = pd.concat([customer, customer_dummies], axis=1)
```

```python
# Add dummies to Segment.io
segment_dummies = pd.get_dummies(segment[['event',
                                          'type',
                                          'traits_isDependent',
                                          'properties_order_type']], 
                                  prefix='dummy', prefix_sep='_')

segment_io = pd.concat([segment, segment_dummies], axis=1)
```

```python
# Combine Segment_io & Customer_io dataframes into df
df = pd.concat([customer_io, segment_io], 
          axis=0, 
          join='outer', 
          ignore_index=True).reset_index(drop=True)
```

```python
# Combine customer.io and segment.io userIds
def concat(*args):
    strs = [str(arg) for arg in args if not pd.isnull(arg)]
    return ','.join(strs) if strs else np.nan
np_concat = np.vectorize(concat)

df['external_member_id'] = np_concat(df['data_customerId'], df['userId'])
```

```python
# Convert UNIX timestamp to datetime
df['datetime'] = pd.to_datetime(df['timestamp'].astype(int), unit='s')
df['date'] = pd.DatetimeIndex(pd.to_datetime(df['datetime'])).date
df['time'] = pd.DatetimeIndex(pd.to_datetime(df['datetime'])).time
df['year'] = pd.DatetimeIndex(pd.to_datetime(df['datetime'])).year
df['month'] = pd.DatetimeIndex(pd.to_datetime(df['datetime'])).month
df['day'] = pd.DatetimeIndex(pd.to_datetime(df['datetime'])).day
df['weekday'] = pd.DatetimeIndex(pd.to_datetime(df['datetime'])).weekday
df['hour'] = pd.DatetimeIndex(pd.to_datetime(df['datetime'])).hour
# Set index to 'datetime'
df.index = df['datetime']
df = df.drop('datetime', axis=1)
```

```python
# Rearrange dataframe columns (so it's easier to read)
pd.set_option('display.max_rows', 500)

df = df[
pd.concat([
pd.Series(df.columns[-pd.Series(df.columns).str.contains('dummy_')]),
pd.Series(df.columns[pd.Series(df.columns).str.contains('dummy_')].order())]).reset_index(drop=True)]
```

```python
events_df = pd.DataFrame(customer.groupby(['data_variables_eventName','timestamp']).apply(len), 
                         columns=['count']).reset_index().sort(columns=['timestamp'])

events_df = pd.DataFrame(events_df.pivot_table(index=['timestamp'], 
                        columns =['data_variables_eventName'], aggfunc=np.sum, fill_value=0))
```

```python
events_cum_df = pd.DataFrame(customer.groupby(['data_variables_eventName','timestamp']).apply(len), 
                         columns=['count']).reset_index().sort(columns=['timestamp'])

events_cum_df = pd.DataFrame(events_cum_df.pivot_table(index=['timestamp'], 
                        columns =['data_variables_eventName'], aggfunc=np.sum, fill_value=0)).cumsum()
```
