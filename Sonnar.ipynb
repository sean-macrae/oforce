```python
import datetime
import itertools
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
from pandas import DataFrame, Series
from pandas.io import gbq
from sklearn import ensemble
from sklearn.utils import shuffle
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import mean_squared_error, explained_variance_score, r2_score
```

```python
###############################################################################
# Query AdWords.ad from Google BigQuery API

projectid = '1014050281203'
query = """SELECT
  timestamp(Date) AS Date,
  INTEGER(STRFTIME_UTC_USEC(Date,'%Y')) AS DateYear,
  INTEGER(STRFTIME_UTC_USEC(Date,'%W')) AS DateWeek,
  CampaignName,
  REGEXP_EXTRACT(CreativeDestinationUrl,
r'(?:utm_campaign|subid)=([a-z0-9-{}\_\|]+)(?:~|&|$)') AS Subid,
  REGEXP_EXTRACT(CampaignName,
r'^[\w\s\-+]*\|[\w\s\-+]*\|\s([\w\s\-+]*)\s\|') AS CountryCode,
  AdType,
  SUM(Impressions) AS Impressions,
  SUM(Clicks) AS Clicks,
  SUM(FLOAT(Cost)/1000000) AS Cost,
  SUM(Conversions) AS Conversions
  FROM
  AdWords.ad
  WHERE
  timestamp(Date) >= TIMESTAMP("2013-01-01")
  GROUP EACH BY
  1,2,3,4,5,6,7"""
    
adwords_ad = pd.read_gbq(query, project_id = projectid, reauth=False)
adwords_ad.head()
```

```python
###############################################################################
# Classify AdWords subchannel via 'AdType' and 'CampaignName' fields

### Create Subchannel column
adwords_ad['Subchannel'] = 'Unknown'

# Classify GDN based on AdType
adwords_ad.loc[
(adwords_ad['AdType'] == 'Display ad') | (adwords_ad['AdType'] == 'Image ad'), 'Subchannel'] = 'House Ads: GDN'

# Identify search based on AdType
adwords_ad.loc[
(adwords_ad['AdType'] == 'Text ad'), 'Subchannel'] = 'House Ads: Search'

# Overwrite records based on CampaignName
adwords_ad.loc[
(adwords_ad['CampaignName'].str.contains('GDN')) | (adwords_ad['CampaignName'].str.contains('gdn')), 'Subchannel']= 'House Ads: GDN'

# Overwrite records based on Subid
adwords_ad['Subid'] = adwords_ad['Subid'].fillna('Unknown')

adwords_ad.loc[
(adwords_ad['Subid'].str.contains('gdn-aw')), 'Subchannel'] = 'House Ads: GDN'
```

```python
###############################################################################
# Query MMR.bigad_country_mapping from Google BigQuery API

projectid = '1014050281203'
query = """SELECT
country_code2 AS CountryCode, 
country_name AS CountryName,
region AS Region,
region2 AS Region2,
subregion AS Subregion,
subregion2 AS Subregion2
FROM MMR.bigad_country_mapping"""
    
mmr_country_map = pd.read_gbq(query, project_id = projectid, reauth=False)
mmr_country_map.head()
```

```python
###############################################################################
# Combine AdWords.ad and MMR Country mapping and aggregate data

# Merge dataframes
adwords_ad = pd.merge(adwords_ad, mmr_country_map, left_on='CountryCode', right_on='CountryCode', how='left')
adwords_ad.head()

# Aggregate to forecast level
adwords_ad = adwords_ad.groupby(['Date','DateYear','DateWeek','CountryName','CountryCode','Subchannel','region','subregion','Subid'])['Impressions','Clicks','Cost','Conversions'].sum()
adwords_ad = adwords_ad.reset_index()

adwords_ad.head()
```

```python
###############################################################################
# Query MMR.Accounts from Google BigQuery API

projectid = '1014050281203'
query = """SELECT
  sups.date AS Date,
  sups.subid AS Subid,
  sups.country AS CountryName,
  sups.subchannel AS Subchannel,
  sups.signUps AS SignUps,
  acq.acquisitions AS Acquisitions
FROM (
  SELECT
    TIMESTAMP(created_date) AS date,
    subid,
    country,
    subchannel,
    COUNT(DISTINCT uaid) AS signups
  FROM
    MMR.Accounts
  WHERE
    TIMESTAMP(created_date) >= TIMESTAMP("2013-01-01")
    GROUP EACH BY 1,
    2,
    3,
    4
    ) sups
  LEFT
JOIN
  EACH
  (
  SELECT
    TIMESTAMP(activated_date) AS date,
    subid,
    country,
    subchannel,
    COUNT(DISTINCT uaid) AS acquisitions
  FROM
    MMR.Accounts
  WHERE
    activated_date IS NOT NULL
    AND TIMESTAMP(activated_date) >= TIMESTAMP("2013-01-01")
    GROUP EACH BY 1,
    2,
    3,
    4
    ) acq
ON
  sups.date = acq.date
  AND sups.subid = acq.subid
  AND sups.country = acq.country
  AND sups.subchannel = acq.subchannel
WHERE
  ((sups.subchannel = "House Ads: Search")
    OR (sups.subchannel = "House Ads: GDN"))"""
    
mmr_accounts = pd.read_gbq(query, project_id = projectid, reauth=False)
mmr_accounts
```

```python
###############################################################################
# Merge MMR and AdWords.ad

# Set dataframe indices to merge on
raw_df = pd.merge(adwords_ad, mmr_accounts, on=['Date','CountryName','Subid'], how='left')

raw_df.describe()
raw_df.head()
```

```python
# Combine Subchannel fields, defaulting to MMR grouping where available; where not available, use custom definitions.

subchannel = []

for i in range(0, len(raw_df)):
    if (pd.Series(raw_df['Subchannel_y'][i]).isnull()[0]):
        subchannel.append(pd.Series(raw_df['Subchannel_x'][i])[0])
    else:
        subchannel.append(pd.Series(raw_df['Subchannel_y'][i])[0])
        
# Save as pandas Series
Subchannel = pd.Series(subchannel, name = 'Subchannel')

# Add combined Subchannel column
raw_df = pd.concat([raw_df, Subchannel], axis=1)
# Drop old Subchannel columns
raw_df = raw_df.drop(['Subchannel_x', 'Subchannel_y'], axis=1)

raw_df.head()
```

```python
###############################################################################
# Aggregate data to forecast level

raw_df = raw_df.groupby(['Date','DateYear','DateWeek',
                'CountryName','Subchannel','Subid'])[['Impressions','Clicks','Cost',
                                                              'SignUps','Acquisitions','Revenue91','Advertisers']].sum().reset_index()

raw_df.columns = [u'Date', u'DateYear', u'DateWeek', u'Country', u'Subchannel', u'Subid', u'Impressions', u'Clicks', u'Cost', u'SignUps', u'Acquisitions', u'Revenue91', u'Advertisers']
```

```python
###############################################################################
# Pre-process raw data

### Remove NaN's from primary dimensions
raw_df = raw_df[raw_df.Country.notnull()] # Country

raw_df = raw_df[raw_df.Subchannel.notnull()] # Subchannel
raw_df = raw_df[raw_df['Subchannel']!='Unknown'].reset_index(drop=True) # Subchannel

raw_df = raw_df[raw_df.Subid.notnull()] # Subid
raw_df = raw_df[raw_df['Subid']!='Unknown'].reset_index(drop=True) # Subid

### Encode categoric variables as numeric factors
country_encode = LabelEncoder()
subchannel_encode = LabelEncoder()
subid_encode = LabelEncoder()

# Country
country_factor = country_encode.fit(raw_df['Country'])
raw_df['CountryFactor'] = country_factor.transform(raw_df['Country'])
# Subchannel
subchannel_factor = subchannel_encode.fit(raw_df['Subchannel'])
raw_df['SubchannelFactor'] = subchannel_factor.transform(raw_df['Subchannel'])
# Subid
subid_factor = subid_encode.fit(raw_df['Subid'])
raw_df['SubidFactor'] = subid_factor.transform(raw_df['Subid'])

### Calculate CTR, CPC, SignUpRate, AcquisitionRate, AcquisitionRunning, ARPA
raw_df['CTR'] = raw_df['Clicks'] / raw_df['Impressions']
raw_df['CPC'] = raw_df['Cost'] / raw_df['Clicks']
raw_df['SignUpRate'] = raw_df['SignUps'] / raw_df['Clicks']
raw_df['AcquisitionRate'] = raw_df['Acquisitions'] / raw_df['SignUps']
raw_df['ARPA91'] = raw_df['Revenue91'] / raw_df['Advertisers']
```

```python
###############################################################################
# Filter data

### Define filters
DateYear_in = [2013, 2014]
DateWeek_in = range(0,52)
Country_in = ['Argentina','Brazil','Chile','Colombia','Mexico','Peru','Canada','United States']
# Country_in = ['United States']
Subchannel_in = ['House Ads: Search', 'House Ads: GDN']

### Apply filters
df = raw_df[raw_df['DateYear'].isin(DateYear_in)]
df = df[df['DateWeek'].isin(DateWeek_in)]
df = df[df['Country'].isin(Country_in)]
df = df[df['Subchannel'].isin(Subchannel_in)]
```

```python
###############################################################################
# Calculate and append traffic and conversion lags

### Create dataframe to lag on
lag_rng = pd.DataFrame(pd.date_range(df['Date'].min(), periods = len(df['Date'].unique()), freq = "D")) # forecast range
lag_rng['Index'] = ([1]*len(lag_rng))
lag_rng

df_iter = df[['Country','Subchannel','Subid']].drop_duplicates().sort(['Country','Subchannel','Subid']).reset_index(drop=True)
df_iter['Index'] = ([1]*len(df_iter))
df_iter

lag = pd.merge(lag_rng, df_iter, on='Index').drop('Index',axis=1)
lag.columns = ['Date','Country','Subchannel','Subid']
lag = lag.sort(['Country','Subchannel','Subid','Date']).reset_index(drop=True) #sort dataframe

### Merge actual data into lagged template
lag = pd.merge(lag,
df[['Date','Country','Subchannel','Subid','Impressions','Clicks','SignUps','Acquisitions','Advertisers','CPC','ARPA91','Revenue91']],
how='left', on=['Country','Subchannel','Subid','Date'])

### Lag data
for i in range(1,15):
    impressions_column = 'ImpressionsLag' + str(i)
    lag[impressions_column] = lag['Impressions'].shift(i)
    
    clicks_column = 'ClicksLag' + str(i)
    lag[clicks_column] = lag['Clicks'].shift(i)
    
    signups_column = 'SignUpsLag' + str(i)
    lag[signups_column] = lag['SignUps'].shift(i)
    
    acquisitions_column = 'AcquisitionsLag' + str(i)
    lag[acquisitions_column] = lag['Acquisitions'].shift(i)
    
    advertisers_column = 'AdvertisersLag' + str(i)
    lag[advertisers_column] = lag['Advertisers'].shift(i)
    
    cpc_column = 'CPCLag' + str(i)
    lag[cpc_column] = lag['CPC'].shift(i)
    
    arpa91_column = 'ARPA91Lag' + str(i)
    lag[arpa91_column] = lag['ARPA91'].shift(i)
    
    revenue91_column = 'Revenue91Lag' + str(i)
    lag[revenue91_column] = lag['Revenue91'].shift(i)
        
### Merge lagged results back into df
cols_to_use = lag.columns - ['Impressions','Clicks','SignUps','Acquisitions','Advertisers','CPC','ARPA91','Revenue91']
df = pd.merge(df, lag[cols_to_use], how='right', on=['Country','Subchannel','Subid','Date'])
df = df.sort(['Country','Subchannel','Subid','Date']).reset_index(drop=True)

## Populate numeric variables with NaN to 0
df = df.fillna(0)

## Error out where lags cross from one categoric group (Country, Subchannel, Subregion) into another
for i in range(1,15):
    date_min = df['Date'].min()
    impressions_column = 'ImpressionsLag' + str(i)
    clicks_column = 'ClicksLag' + str(i)
    signups_column = 'SignUpsLag' + str(i)
    acquisitions_column = 'AcquisitionsLag' + str(i)
    advertisers_column = 'AdvertisersLag' + str(i)
    cpc_column = 'CPCLag' + str(i)
    arpa91_column = 'ARPA91Lag' + str(i)
    revenue91_column = 'Revenue91Lag' + str(i)
    for j in range(0,i):
        df.loc[df['Date']==(date_min + datetime.timedelta(days=(j+0))),impressions_column] = np.nan
        df.loc[df['Date']==(date_min + datetime.timedelta(days=(j+0))),clicks_column] = np.nan
        df.loc[df['Date']==(date_min + datetime.timedelta(days=(j+0))),signups_column] = np.nan
        df.loc[df['Date']==(date_min + datetime.timedelta(days=(j+0))),acquisitions_column] = np.nan
        df.loc[df['Date']==(date_min + datetime.timedelta(days=(j+0))),advertisers_column] = np.nan
        df.loc[df['Date']==(date_min + datetime.timedelta(days=(j+0))),cpc_column] = np.nan
        df.loc[df['Date']==(date_min + datetime.timedelta(days=(j+0))),arpa91_column] = np.nan
        df.loc[df['Date']==(date_min + datetime.timedelta(days=(j+0))),revenue91_column] = np.nan
```

```python
###############################################################################
# Pre-process calendar lag data

### Append calendar meta-variables (breaks Date into component pieces)
df['DateYear'] = pd.DatetimeIndex(df['Date']).year # Day of Week
df['DateWeek'] = pd.DatetimeIndex(df['Date']).week # Month of Year
df['DateWeekday'] = pd.DatetimeIndex(df['Date']).weekday # Day of Week
df['DateMonth'] = pd.DatetimeIndex(df['Date']).month # Month of Year
df['DateDay'] = pd.DatetimeIndex(df['Date']).day # Day of Month
df['DateDayOfYear'] = pd.DatetimeIndex(df['Date']).dayofyear # Day of Year
df['DateQuarter'] = pd.DatetimeIndex(df['Date']).quarter # Quarter of Year
df['DateMonthEnd'] = pd.DatetimeIndex(df['Date']).is_month_end.astype(int) # Logical flag if end of month
df['DateQuarterEnd'] = pd.DatetimeIndex(df['Date']).is_quarter_end.astype(int) # Logical flag if end of month

### Define calendar events by year (to calculate lags off of)
newyears_dates = [pd.datetime(2012,1,1), pd.datetime(2013,1,1), pd.datetime(2014,1,1), pd.datetime(2015,1,1), pd.datetime(2016,1,1)] # New Years Day
presidents_dates = [pd.datetime(2012,2,20), pd.datetime(2013,2,18), pd.datetime(2014,2,17), pd.datetime(2015,2,16), pd.datetime(2016,2,15)] # Presidents Day
easter_dates = [pd.datetime(2012,4,8), pd.datetime(2013,3,31), pd.datetime(2014,4,20), pd.datetime(2015,4,5), pd.datetime(2016,3,27)] # Easter
aprilfools_dates = [pd.datetime(2012,4,1), pd.datetime(2013,4,1), pd.datetime(2014,4,1), pd.datetime(2015,4,1), pd.datetime(2016,4,1)] # April Fools
thanksgiving_dates = [pd.datetime(2012,11,22), pd.datetime(2013,11,28), pd.datetime(2014,11,27), pd.datetime(2015,11,26), pd.datetime(2016,11,24)] # Thanksgiving
xmas_dates = [pd.datetime(2012,12,25), pd.datetime(2013,12,25), pd.datetime(2014,12,25), pd.datetime(2015,12,25), pd.datetime(2016,12,25)] # Christmas
summer_solstice = [pd.datetime(2012,6,20), pd.datetime(2013,6,20), pd.datetime(2014,6,21), pd.datetime(2015,6,21), pd.datetime(2016,6,20)] # Summer Solstice
autumnal_equinox = [pd.datetime(2012,9,22), pd.datetime(2013,9,22), pd.datetime(2014,9,23), pd.datetime(2015,9,23), pd.datetime(2016,9,22)] # Autumnal Equinox

### Combine all calendar lags into a list with corresponding list of names
lag_dates = [newyears_dates, presidents_dates, easter_dates, 
             aprilfools_dates, thanksgiving_dates, xmas_dates, 
             summer_solstice, autumnal_equinox]
lag_names = ['NewYearsLag', 'PresidentsLag', 'EasterLag',
             'AprilFoolsLag', 'ThanksgivingLag', 'XmasLag', 
             'SummerLag', 'AutumnLag']

### Calculate calendar lags
lags = [[] for x in xrange(len(lag_dates))]
lags

for k in range(0,len(lag_dates)):
    for i in df['Date']:
        for j in range(0,len(lag_dates[k])):
            if (lag_dates[k][j] - i).days < 0:
                pass
            else:
                lags[k].append((lag_dates[k][j] - i).days)
                break

### Append calendar lags to dataframe
for i in range(0,len(lag_names)):
    df[lag_names[i]] = lags[i]
```

```python
###############################################################################
### Outlier analysis and removal

### Define outlier fence function
def outlier_fences(x):
    # subset dataframe on numeric feilds only
    msk = (x.dtypes == 'int64') | (x.dtypes == 'float64')
    x = x.loc[:, msk]
    # quartile distributions
    q1 = x.quantile(q=0.25)
    q3 = x.quantile(q=0.75)
    iq = (q3 - q1)
    # outlier fences
    # lower
    lower_inner = (q1 - (5.0*iq)) # mild
    lower_outer = (q1 - (7.0*iq)) # extreme
    # upper
    upper_inner = (q3 + (5.0*iq)) # mild
    upper_outer = (q3 + (7.0*iq)) # extreme
    fences = [lower_inner,lower_outer,upper_inner,upper_outer]
    return fences

# Original size of dataframe before excluding outliers
df_original_size = df.count()[0]

# Difference series'
df_diff = abs(df[['Impressions','Clicks','SignUps','Acquisitions','Revenue91']].diff(1))
df_diff.columns = [['ImpressionsDiff','ClicksDiff','SignUpsDiff','AcquisitionsDiff','Revenue91Diff']]

df_diff = pd.concat([df[['Date','Country','Subchannel','Subid',
                         'Impressions','Clicks','SignUps','Acquisitions','Revenue91',
                         'ImpressionsMA','ClicksMA','SignUpsMA','AcquisitionsMA','Revenue91MA']],
                     df_diff], axis=1)

# Set first record per forecast group = np.nan
df_diff.loc[df_diff['Date']==df_diff['Date'].min(),
            ['ImpressionsDiff','ClicksDiff','SignUpsDiff','AcquisitionsDiff','Revenue91Diff']] = np.nan

# Caclulate % diff of moving average
df_diff['ImpressionsMA_pct'] = df_diff['ImpressionsDiff'] / df_diff['ImpressionsMA']
df_diff['ClicksMA_pct'] = df_diff['ClicksDiff'] / df_diff['ClicksMA']
df_diff['SignUpsMA_pct'] = df_diff['SignUpsDiff'] / df_diff['SignUpsMA']
df_diff['AcquisitionsMA_pct'] = df_diff['AcquisitionsDiff'] / df_diff['AcquisitionsMA']
df_diff['Revenue91MA_pct'] = df_diff['Revenue91Diff'] / df_diff['Revenue91MA']

# Number of non-NaN's within a moving 15 day moving window
df_diff_nan = pd.rolling_count(
    df_diff[[
    'ImpressionsMA_pct','ClicksMA_pct','SignUpsMA_pct','AcquisitionsMA_pct','Revenue91MA_pct']], window = 15, center=True)

df_diff_nan.columns = ['ImpressionsMA_nan','ClicksMA_nan','SignUpsMA_nan','AcquisitionsMA_nan','Revenue91MA_nan']
df_diff = pd.concat([df_diff, df_diff_nan], axis=1)

# Filter dataframe for stable segments and observations/moving average based on Impressions
df_diff = df_diff[
(df_diff['ImpressionsMA_nan'] > 12)]

group = df_diff[[
'Country','Subchannel','Subid',
'ImpressionsMA_pct','ClicksMA_pct']].groupby(by = ['Country','Subchannel','Subid'])

df_outliers = group.apply(outlier_fences)

### Calculate outliers ranges

# Create first record in dataframe
disjoint_window = pd.DataFrame(pd.DataFrame(df_outliers[0]).unstack()).rename(columns = {0:'Fence'}, index = {0:'LowerInner', 1:'LowerOuter', 2:'UpperInner', 3:'UpperOuter'}).transpose()
test_index = df_outliers.reset_index()[['Country','Subchannel','Subid']][0:1].rename(index = {0:'Fence'})
outliers = pd.concat([disjoint_window, test_index], axis=1).reset_index(drop=True)
    
# Append additional records
for i in range(1, len(df_outliers)):
    disjoint_window = pd.DataFrame(pd.DataFrame(df_outliers[i]).unstack()).rename(columns = {0:'Fence'}, index = {0:'LowerInner', 1:'LowerOuter', 2:'UpperInner', 3:'UpperOuter'}).transpose()
    test_index = df_outliers.reset_index()[['Country','Subchannel','Subid']][i:(i+1)].rename(index = {i:'Fence'})
    test_append = pd.concat([disjoint_window, test_index], axis=1).reset_index(drop=True)
    outliers = pd.concat([outliers, test_append], axis=0, ignore_index=True)
    
outliers = outliers.set_index(['Country','Subchannel','Subid'])

# Merge outliers with df_diff dataframe
df_diff = df_diff.set_index(['Country','Subchannel','Subid']).join(outliers)
df_diff = df_diff.reset_index()

# Flag outliers
df_diff['ImpressionsOutlier'] = (pd.Series(np.round(df_diff['ImpressionsMA_pct'], 6)) <= 
                                 pd.Series(np.round(df_diff[('ImpressionsMA_pct', 'LowerOuter')], 6))) | (pd.Series(np.round(df_diff['ImpressionsMA_pct'], 6)) >= 
                                                                                                          pd.Series(np.round(df_diff[('ImpressionsMA_pct', 'UpperOuter')], 6)))

# Exclude outliers from dataframe
df_outliers = df_diff[df_diff['ImpressionsOutlier'] == 1][['Date','Country','Subchannel','Subid','ImpressionsOutlier']]
df = df[df.merge(df_outliers, how='left')['ImpressionsOutlier'].isnull()]
```

```python
###############################################################################
# Create forecast dataframe

### Calculate forecast start and end dates
forecast_start = df['Date'].max() + datetime.timedelta(days=1)
forecast_end = pd.date_range(start = df['Date'].max(), periods=2, freq='Q').max()

# Create forecast range
forecast_rng = pd.date_range(start=forecast_start, end=forecast_end, freq='D')
forecast_len = len(forecast_rng)

forecast = []

rowlists = [forecast_rng, Country_in, Subchannel_in, Branding_in]
for combination in itertools.product(*rowlists):
    forecast.append(combination)
    
forecast = pd.DataFrame(forecast)
forecast.columns = ['Date','Country','Subchannel','Branding']

# Encode categoric variables
forecast['CountryFactor'] = country_factor.transform(forecast['Country'])
forecast['SubchannelFactor'] = subchannel_factor.transform(forecast['Subchannel'])
forecast['BrandingFactor'] = branding_factor.transform(forecast['Branding'])

### Append meta-calendar variables
forecast['DateYear'] = pd.DatetimeIndex(forecast['Date']).year # Year
forecast['DateWeek'] = pd.DatetimeIndex(forecast['Date']).week # Week of Year
forecast['DateWeekday'] = pd.DatetimeIndex(forecast['Date']).weekday # Day of Week
forecast['DateMonth'] = pd.DatetimeIndex(forecast['Date']).month # Month of Year
forecast['DateWeekday'] = pd.DatetimeIndex(forecast['Date']).weekday # Day of Week
forecast['DateMonth'] = pd.DatetimeIndex(forecast['Date']).month # Month of Year
forecast['DateDay'] = pd.DatetimeIndex(forecast['Date']).day # Day of Month
forecast['DateDayOfYear'] = pd.DatetimeIndex(forecast['Date']).dayofyear # Day of Year
forecast['DateQuarter'] = pd.DatetimeIndex(forecast['Date']).quarter # Quarter of Year
forecast['DateMonthEnd'] = pd.DatetimeIndex(forecast['Date']).is_month_end.astype(int) # Logical flag if end of month
forecast['DateQuarterEnd'] = pd.DatetimeIndex(forecast['Date']).is_quarter_end.astype(int) # Logical flag if end of month

### Create calendar lags for forecast dates
lags = [[] for x in xrange(len(lag_dates))]
lags

for k in range(0,len(lag_dates)):
    for i in forecast['Date']:
        for j in range(0,len(lag_dates[k])):
            if (lag_dates[k][j] - i).days < 0:
                pass
            else:
                lags[k].append((lag_dates[k][j] - i).days)
                break

### Append calendar lags to forecast dataframe
for i in range(0,len(lag_names)):
    forecast[lag_names[i]] = lags[i]
    

### Append blank traffic and conversion fields
forecast['Impressions'] = np.zeros(len(forecast['Date']))
forecast['Clicks'] = np.zeros(len(forecast['Date']))
forecast['SignUps'] = np.zeros(len(forecast['Date']))
forecast['Acquisitions'] = np.zeros(len(forecast['Date']))
forecast['Advertisers'] = np.zeros(len(forecast['Date']))
forecast['CPC'] = np.zeros(len(forecast['Date']))
forecast['ARPA91'] = np.zeros(len(forecast['Date']))
forecast['Revenue91'] = np.zeros(len(forecast['Date']))
forecast['Revenue365'] = np.zeros(len(forecast['Date']))

### Append average of last 5 days as trend assumptions moving forward (flat-line)

def last_30(x):
    return x[-30:]

# Select tail of trends by forecast dimensions
forecast_trend = df[['Country','Subchannel','Branding','ImpressionsTrendYear','ImpressionsTrendDay','ImpressionsMA',
    'ClicksTrendYear','ClicksTrendDay','ClicksMA','SignUpsTrendYear','SignUpsTrendDay','SignUpsMA','AcquisitionsTrendYear','AcquisitionsTrendDay','AcquisitionsMA',
    'AdvertisersTrendYear', 'AdvertisersTrendDay','AdvertisersMA','CPCTrendYear', 'CPCTrendDay','CPCMA', 'ARPA91TrendYear', 'ARPA91TrendDay','ARPA91MA',
    'Revenue91TrendYear','Revenue91TrendDay','Revenue365TrendYear','Revenue365TrendDay']].dropna().groupby(['Country','Subchannel','Branding']).apply(last_30)

# Calculate average of tail of trends by forecast dimensions
forecast_trend = forecast_trend[['Country','Subchannel','Branding','ImpressionsTrendYear','ImpressionsTrendDay','ImpressionsMA',
    'ClicksTrendYear','ClicksTrendDay','ClicksMA','SignUpsTrendYear','SignUpsTrendDay','SignUpsMA','AcquisitionsTrendYear','AcquisitionsTrendDay','AcquisitionsMA',
    'AdvertisersTrendYear', 'AdvertisersTrendDay','AdvertisersMA','CPCTrendYear', 'CPCTrendDay','CPCMA', 'ARPA91TrendYear', 'ARPA91TrendDay','ARPA91MA',
    'Revenue91TrendYear','Revenue91TrendDay','Revenue365TrendYear','Revenue365TrendDay']].groupby(['Country','Subchannel','Branding']).mean()

# Merge trend assumptions into forecast dataframe
forecast = forecast.merge(forecast_trend.reset_index(), how='left', on=['Country','Subchannel','Branding'])

### Append historic data and call the new dataframe Sonnar
sonnar = df[forecast.columns].append(forecast, ignore_index=True).sort(columns=['Country','Subchannel','Branding','Date'])

# update historical trends where most recent records are NaN (due to centered moving average)
# NOT WORKING
# sonnar[(sonnar['Date'] > test_date_end) & 
#            (np.isnan(sonnar['ImpressionsTrendYear']))].set_index(['Date','Country','Subchannel','Branding']).update(forecast_trend, overwrite=True)

sonnar = sonnar.reset_index()

### Calculate traffic and conversion lags to forecast dataframe
for i in range(1,15):
    impressions_column = 'ImpressionsLag' + str(i)
    sonnar[impressions_column] = sonnar['Impressions'].shift(i)
    
    clicks_column = 'ClicksLag' + str(i)
    sonnar[clicks_column] = sonnar['Clicks'].shift(i)
    
    signups_column = 'SignUpsLag' + str(i)
    sonnar[signups_column] = sonnar['SignUps'].shift(i)
    
    acquisitions_column = 'AcquisitionsLag' + str(i)
    sonnar[acquisitions_column] = sonnar['Acquisitions'].shift(i)
    
    advertisers_column = 'AdvertisersLag' + str(i)
    sonnar[advertisers_column] = sonnar['Advertisers'].shift(i)
    
    cpc_column = 'CPCLag' + str(i)
    sonnar[cpc_column] = sonnar['CPC'].shift(i)
    
    arpa91_column = 'ARPA91Lag' + str(i)
    sonnar[arpa91_column] = sonnar['ARPA91'].shift(i)
    
    revenue91_column = 'Revenue91Lag' + str(i)
    sonnar[revenue91_column] = sonnar['Revenue91'].shift(i)
    
    revenue365_column = 'Revenue365Lag' + str(i)
    sonnar[revenue365_column] = sonnar['Revenue365'].shift(i)   
    
## Set multi-index on Sonnar for merging predictions in iterative prediction method below
sonnar = sonnar.set_index(['Date','Country','Subchannel','Branding'])

# Replace infinity with NaN
sonnar = sonnar.replace([np.inf, -np.inf], np.nan)

## Populate numeric variables with NaN to 0
sonnar = sonnar.fillna(0)
```

```python
############################################################################### 
# Filter GBRT: Iterative GBM Forecast

### Forecast 
for i in range(1,(forecast_len+1)): 
    predict_day = df['Date'].max() + datetime.timedelta(days=i) # day to predict 
    predict_set = sonnar.filter(like=str(predict_day.date()), axis = 0) # prediction set
    # Impressions prediction
    impressions_predict = pd.Series(np.round(gbm_impressions.predict(predict_set[impressions_signals]), 0), name='Impressions')
    impressions_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    impressions_index = impressions_index.join(impressions_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(impressions_index)
    # Clicks prediction
    clicks_predict = pd.Series(np.round(gbm_clicks.predict(predict_set[clicks_signals]), 0), name='Clicks')
    clicks_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    clicks_index = clicks_index.join(clicks_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(clicks_index)
    # Signups prediction
    signups_predict = pd.Series(np.round(gbm_signups.predict(predict_set[signups_signals]), 0), name='SignUps')
    signups_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    signups_index = signups_index.join(signups_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(signups_index)
    # Acquisitions prediction
    acquisitions_predict = pd.Series(np.round(gbm_acquisitions.predict(predict_set[acquisitions_signals]), 0), name='Acquisitions')
    acquisitions_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    acquisitions_index = acquisitions_index.join(acquisitions_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(acquisitions_index)
    # Advertisers prediction
    advertisers_predict = pd.Series(np.round(gbm_advertisers.predict(predict_set[advertisers_signals]), 0), name='Advertisers')
    advertisers_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    advertisers_index = advertisers_index.join(advertisers_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(advertisers_index)
    # CPC prediction
    cpc_predict = pd.Series(np.round(gbm_cpc.predict(predict_set[cpc_signals]), 0), name='CPC')
    cpc_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    cpc_index = cpc_index.join(cpc_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(cpc_index)
    # ARPA91 prediction
    arpa91_predict = pd.Series(np.round(gbm_arpa91.predict(predict_set[arpa91_signals]), 0), name='ARPA91')
    arpa91_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    arpa91_index = arpa91_index.join(arpa91_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(arpa91_index)
    # Revenue91 prediction
    revenue91_predict = pd.Series(np.round(gbm_revenue91.predict(predict_set[revenue91_signals]), 0), name='Revenue91')
    revenue91_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    revenue91_index = revenue91_index.join(revenue91_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(revenue91_index)
    # Revenue365 prediction
    revenue365_predict = pd.Series(np.round(gbm_revenue365.predict(predict_set[revenue365_signals]), 0), name='Revenue365')
    revenue365_index = pd.DataFrame(predict_set.reset_index()[['Date','Country','Subchannel','Branding']])
    revenue365_index = revenue365_index.join(revenue365_predict).set_index(['Date','Country','Subchannel','Branding'])
    predict_set.update(revenue365_index)
    
    # merge predictions with Sonnar
    predict = predict_set[['Impressions','Clicks','SignUps','Acquisitions','Advertisers','CPC','ARPA91','Revenue91','Revenue365']]
    sonnar.update(predict) # update Sonnar with predictions
    
    for k in range(1,15): # update prediction lags 1 - 14 
        # create lag dataframe with multindex and update Sonnar
        impressions_column = 'ImpressionsLag' + str(k)
        clicks_column = 'ClicksLag' + str(k)
        signups_column = 'SignUpsLag' + str(k)
        acquisitions_column = 'AcquisitionsLag' + str(k)
        advertisers_column = 'AdvertisersLag' + str(k)
        cpc_column = 'CPCLag' + str(k)
        arpa91_column = 'ARPA91Lag' + str(k)
        revenue91_column = 'Revenue91Lag' + str(k)
        lag = pd.DataFrame([list(predict_set.reset_index()['Country']),
            list(predict_set.reset_index()['Subchannel']), 
            list(predict_set.reset_index()['Branding']),
            list(predict_set.reset_index()['Date'] + datetime.timedelta(days=k)),
            list(impressions_predict),
            list(clicks_predict),
            list(signups_predict),
            list(acquisitions_predict),
            list(advertisers_predict),
            list(cpc_predict),
            list(arpa91_predict),
            list(revenue91_predict),
            list(revenue365_predict)]).transpose().rename(
            columns = {0:'Country', 1:'Subchannel', 2:'Branding', 3:'Date', 4:impressions_column, 5:clicks_column, 
                       6:signups_column, 7:acquisitions_column, 8:advertisers_column, 9:cpc_column, 10:arpa91_column, 11:revenue91_column, 12:revenue365_column}).set_index(['Date','Country','Subchannel','Branding']) # make prediction 
        sonnar.update(lag) # update Sonnar with prediction lags
```
