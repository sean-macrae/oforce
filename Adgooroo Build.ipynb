```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import os
```

```python
# get list of files and folders in Adgooroo directory
path = r'I:\AKQA_Search\Clients\Intuit\TurboTax\Reporting\Weekly Competitive Reports\AdGooRoo Data'
folderlist = os.listdir(path)
```

```python
folderlist
```

```python
# create list of strings for the directory path of each files in Adgooroo directory
data = []
for dir_entry in os.listdir(path):
    dir_entry_path = os.path.join(path, dir_entry)
    if os.path.isfile(dir_entry_path):
        data.append(dir_entry_path)
```

```python
data
```

```python
# write 'Sheet1' of each file into a list of pandas dataframes
pieces = []

for i in data:
    workbook = pd.ExcelFile(i)
    worksheet = workbook.parse('Sheet1')
    
    worksheet['Directory'] = i
    pieces.append(worksheet)
```

```python
# concatenate list of all dataframes into one dataframe
adgooroo = pd.concat(pieces, ignore_index=True)
```

```python
# define function to return count of records
def f(x):
    return Series([len(x.index)], index=['Records'])
```

```python
# group data by directory path and apply count function, f
directory_groups = adgooroo.groupby(adgooroo['Directory'])
directory_stats = directory_groups.apply(f)
```

```python
# head of the count of records by file, sorted descending
directory_stats.sort(['Records'],ascending=False)[0:10] # check for outliers, anomolies
```

```python
# replace NA's with 0's
adgooroo = adgooroo.fillna(0)
```

```python
# define function to calculate summary stats over dataframe
def g(x):
    return Series([len(x.index),x.min(),x.max(),x.mean(),x.median()], index=['Records','Min','Max','Mean','Median'])
```

```python
# group dataframe by Advertiser and apply summary stats to their Coverage
advertiser_groups = adgooroo.groupby(adgooroo['Advertiser'])
advertiser_stats = advertiser_groups['Coverage'].apply(g)
```

```python
# unstack advertiser stats
advertiser_to_map = advertiser_stats.unstack()
```

```python
# write advertiser aggregation to csv for manual mapping
advertiser_to_map.to_csv('I:\AKQA_Search\Clients\Intuit\TurboTax\Reporting\Weekly Competitive Reports\AdGooRoo Data\Advertiser Map\\advertiser_to_map.csv', encoding='utf-8')
```
