```python
import numpy as np
import scipy.stats
import statsmodels.stats.power as smp
import pandas as pd
from pandas import DataFrame, Series
from pandas.io import gbq
import matplotlib.pyplot as plt
```

```python
###############################################################################
# Load data from Google BigQuery API

projectid = '1014050281203'
query = """SELECT
CampaignName, cid, CreativeDestinationUrl, Date, 
SUM(Impressions) AS Impressions,
SUM(Clicks) AS Clicks,
SUM(Conversions) AS Conversions,
AVG(BounceRate) AS BounceRate
FROM
AdWords.ad
WHERE
cid = '10015085'
GROUP EACH BY 1,2,3,4"""
    
raw_df = pd.read_gbq(query, project_id = projectid)
```

```python
###############################################################################
# Calculate CTR

raw_df['CTR'] = raw_df['Clicks'] / raw_df['Impressions']
```

```python
###############################################################################
# Extract Subid

subid_proc = raw_df['CreativeDestinationUrl'].str.split('subid=').tolist()
subid_proc = pd.DataFrame(subid_proc)[1]
subid_proc = subid_proc.str.split('~').tolist()

subid = []

for i in range(0,len(subid_proc)):
    if subid_proc[i] == None: 
        subid.append('None')    
    else: 
        subid.append(subid_proc[i][0])
        
raw_df['Subid'] = pd.DataFrame(subid)
```

```python
###############################################################################
# Print Raw Data Summary Metrics

print "Campaign Name:"
print raw_df['CampaignName'].unique()
print "\n"

print "CID:"
print raw_df['cid'].unique()
print "\n"

print "Raw Data Date Description:"
print pd.Series(raw_df['Date']).describe()
print "\n"

print "Raw Data Description:"
print raw_df.describe()
print "\n"
```

```python
###############################################################################
# Analysis on Impressions per Day
traffic_ts = raw_df.groupby(['Date'])['Impressions','Clicks'].sum()[:-1]
traffic_ts['CTR'] = traffic_ts['Clicks'] / traffic_ts['Impressions']

# Ask Ara about dates w/ no traffic (CID paused?)
print 'Impressions:', np.round(traffic_ts['Impressions'][-30:].mean(),0)
print 'Clicks:', np.round(traffic_ts['Clicks'][-30:].mean(),0)
print 'CTR:', np.round(traffic_ts['CTR'][-30:].mean(),4)
```

```python
###############################################################################
### A/B Dichotomous Sample Size Estimate
# Chi^2 Contingency Table Test for Independence
# kwargs:
# 1) effect_size = (control / treatment) - 1
# 2) alpha = Type I error
# 3) power = 1 minus Type II error
# 4) n_bins = degrees of freedom = (r - 1)(c - 1) + 1 for r = rows and c = columns in contingency table

effect_sizes = [0.001,0.002,0.003,0.004,0.005,0.006,0.007,0.008,0.009,0.01]
sample_sizes = []

for i in effect_sizes:
    sample_sizes.append(round(smp.GofChisquarePower().solve_power(effect_size=i, alpha=0.05, power=0.8, n_bins=(2-1)*(2-1)+1),0))
    
test_times = pd.DataFrame(sample_sizes,effect_sizes)
test_times.reset_index(inplace=True)
test_times.columns = ['effect_sizes','n_obs']

print test_times
```

```python
###############################################################################
### Create traffic to test dataframe

# avg_impressions = traffic_ts['Impressions'][-30:].mean()
avg_impressions = 5200

traffic_alloc = range(90,112,2)

for i in range(0,len(traffic_alloc)):
    traffic_alloc[i] = traffic_alloc[i]/100.0
    
traffic = []

for i in range(0,len(traffic_alloc)):
    traffic.append(round(traffic_alloc[i] * avg_impressions,0))
    
traffic = pd.DataFrame(traffic, traffic_alloc).transpose()
traffic_to_test = traffic

for i in range(0, len(effect_sizes)-1):
    traffic_to_test = traffic_to_test.append(traffic)

traffic_to_test.reset_index(drop=True, inplace=True)

print avg_impressions
```

```python
###############################################################################
### Combine Test Times & Traffic to Test DataFrames

test_times = test_times.join(traffic_to_test, how='inner')

# Calculate Number of Days to Reach Statistical Significance

for i in traffic_alloc:
    for j in range(0,len(effect_sizes)):
        test_times[i][j] = np.ceil(test_times['n_obs'][j] / test_times[i][j])
        
test_times
```

```python
###############################################################################
### Analysis on Effect Size

# ## Subset data on control and experiment groups
# # Subset control
# control = raw_df[((raw_df['Subid'] == 'ca-en-ha-aw-autoc') | (raw_df['Subid'] == 'ca-en-ha-aw-autocm'))]
# print control.describe()
# # Subset experiment
# experiment = raw_df[((raw_df['Subid'] == 'ca-en-ha-aw-autoe') | (raw_df['Subid'] == 'ca-en-ha-aw-autoem'))]
# print "\n", experiment.describe()

## Subset data on control and experiment groups
# Subset control
control = raw_df[((raw_df['Subid'] == 'ca-en-ha-aw-autocm'))]
print control.describe()
# Subset experiment
experiment = raw_df[((raw_df['Subid'] == 'ca-en-ha-aw-autoem'))]
print "\n", experiment.describe()

# # Test for
# control = raw_df[(raw_df['Subid'] == 'ca-en-gdn-gdnsuc')]
# experiment = raw_df[(raw_df['Subid'] == 'ca-en-gdn-dcosuc')]

### Create contingency table
control_clicks = control['Clicks'].sum()
control_nonclicks = control['Impressions'].sum() - control['Clicks'].sum()

experiment_clicks = experiment['Clicks'].sum()
experiment_nonclicks = experiment['Impressions'].sum() - experiment['Clicks'].sum()

clicks = [control_clicks, experiment_clicks]
nonclicks = [control_nonclicks, experiment_nonclicks]

cont_table = np.array([clicks, nonclicks])
print control_clicks
print control_nonclicks
print experiment_clicks
print experiment_nonclicks
```

```python
N = control['Impressions'].sum()
chi2, p, dof, expected = scipy.stats.chi2_contingency(cont_table)
phi = np.sqrt(chi2/N)
print "Chi2, p-value, DoF, Phi (Effect Size):"
print chi2, p, dof, phi
print 'p =', p, "\n"

print pd.DataFrame(cont_table).rename(columns = {0:'Standard', 1:'Automobile'}, index = {0:'Clicks', 1:'Non-Clicks'})
```
